```python
import re

# Funo para extrair informaes do texto
def extract_info(text):
    # Expresses regulares para extrair ns e conexes
    node_pattern = r'(\d+)\sDefaultNodeGene\(key=(\d+),\sbias=(-?\d+\.\d+),\sresponse=(-?\d+\.\d+),\sactivation=(\w+),\saggregation=(\w+)\)'
    connection_pattern = r'DefaultConnectionGene\(key=\(([-?\d+,\s]+)\),\sweight=(-?\d+\.\d+),\senabled=(\w+)\)'

    # Procura por correspondncias nos padres de n e conexo
    nodes = re.findall(node_pattern, text)
    connections = re.findall(connection_pattern, text)

    # Processa ns e converte para formato desejado
    processed_nodes = {}
    for node in nodes:
        key = int(node[0])
        values = {
            'bias': float(node[2]),
            'response': float(node[3]),
            'activation': node[4],
            'aggregation': node[5],
            'connections': []
        }
        processed_nodes[key] = values

    # Processa conexes e converte para formato desejado
    processed_connections = []
    for connection in connections:
        key = tuple(map(int, connection[0].split(',')))
        values = {
            'weight': float(connection[1]),
            'enabled': connection[2] == 'True'
        }
        processed_connections.append((key, values))

    # Adiciona as conexes a cada n
    for connection in processed_connections:
        from_node, to_node = connection[0]
        if from_node >= 0 and connection[1]['enabled']:
            processed_nodes[to_node]['connections'].append(connection)

    return processed_nodes

# Texto fornecido
texto = '''
Nodes:
    0 DefaultNodeGene(key=0, bias=1.7653346324454544, response=1.0677440253475865, activation=sigmoid, aggregation=sum)
    1 DefaultNodeGene(key=1, bias=2.104094178281087, response=0.969113598790605, activation=sigmoid, aggregation=sum)
    2 DefaultNodeGene(key=2, bias=-0.19316363519456758, response=1.0059432907080101, activation=sigmoid, aggregation=sum)
    3 DefaultNodeGene(key=3, bias=-0.7393031089861353, response=0.9166646933367428, activation=sigmoid, aggregation=max)
    4 DefaultNodeGene(key=4, bias=-0.6407311430784627, response=1.0255214818000693, activation=gauss, aggregation=product)
    5 DefaultNodeGene(key=5, bias=4.812240524275377, response=1.0417577554159483, activation=gauss, aggregation=sum)
    6 DefaultNodeGene(key=6, bias=1.277229530307713, response=0.8598038414367637, activation=gauss, aggregation=max)
    7 DefaultNodeGene(key=7, bias=2.2442612110541815, response=1.1249628318053708, activation=sigmoid, aggregation=maxabs)
    8 DefaultNodeGene(key=8, bias=2.3433672937067915, response=1.068286881612852, activation=sigmoid, aggregation=maxabs)
    9 DefaultNodeGene(key=9, bias=-1.6080055259612607, response=0.7963567227110464, activation=sigmoid, aggregation=product)
    10 DefaultNodeGene(key=10, bias=-1.1555293684410215, response=0.8116932683735931, activation=gauss, aggregation=mean)
    11 DefaultNodeGene(key=11, bias=-0.31816165532098095, response=1.0312011983954517, activation=sigmoid, aggregation=mean)
    30 DefaultNodeGene(key=30, bias=0.19788465105857794, response=0.8216185974778835, activation=gauss, aggregation=mean)
    209 DefaultNodeGene(key=209, bias=-0.28732153011409634, response=0.8604623556211811, activation=sigmoid, aggregation=median)
    260 DefaultNodeGene(key=260, bias=0.33882674930694306, response=1.0435286532759804, activation=sigmoid, aggregation=product)
    652 DefaultNodeGene(key=652, bias=0.9327788158170934, response=0.946875840452248, activation=sigmoid, aggregation=mean)
    879 DefaultNodeGene(key=879, bias=2.616427388753264, response=1.0802335701268193, activation=sigmoid, aggregation=product)
    1294 DefaultNodeGene(key=1294, bias=-0.1743452555046639, response=1.0056645894426317, activation=sigmoid, aggregation=maxabs)
    1352 DefaultNodeGene(key=1352, bias=1.693946518030224, response=0.9545467864186808, activation=gauss, aggregation=max)
    1402 DefaultNodeGene(key=1402, bias=-0.17137745794082065, response=0.9787845489343964, activation=sigmoid, aggregation=maxabs)
Connections:
    DefaultConnectionGene(key=(-169, 3), weight=0.04101156560847516, enabled=False)
	DefaultConnectionGene(key=(-169, 6), weight=0.6415383763101508, enabled=False)
	DefaultConnectionGene(key=(-169, 8), weight=3.7547253753842864, enabled=False)
	DefaultConnectionGene(key=(-169, 9), weight=1.3141242555453532, enabled=False)
	DefaultConnectionGene(key=(-168, 0), weight=2.059007032698155, enabled=True)
	DefaultConnectionGene(key=(-168, 4), weight=1.1248835773665284, enabled=False)
	DefaultConnectionGene(key=(-168, 7), weight=0.20151118549298036, enabled=False)
	DefaultConnectionGene(key=(-168, 8), weight=-3.873599641909331, enabled=False)
	DefaultConnectionGene(key=(-168, 11), weight=-0.5469774226076248, enabled=True)
	DefaultConnectionGene(key=(-167, 0), weight=-0.025267029243647654, enabled=False)
	DefaultConnectionGene(key=(-167, 4), weight=-0.639552648415271, enabled=True)
	DefaultConnectionGene(key=(-167, 6), weight=-0.16505312092066848, enabled=False)
	DefaultConnectionGene(key=(-167, 9), weight=0.48447265022774877, enabled=False)
	DefaultConnectionGene(key=(-167, 10), weight=-1.2992151460898151, enabled=False)
	DefaultConnectionGene(key=(-167, 11), weight=-0.2769118217797468, enabled=False)
	DefaultConnectionGene(key=(-166, 0), weight=2.58659059515438, enabled=False)
	DefaultConnectionGene(key=(-166, 1), weight=-0.316246001512942, enabled=True)
	DefaultConnectionGene(key=(-166, 2), weight=-2.326167473283251, enabled=True)
	DefaultConnectionGene(key=(-166, 4), weight=0.2781012185595586, enabled=True)
	DefaultConnectionGene(key=(-166, 7), weight=-3.0371717880647924, enabled=True)
	DefaultConnectionGene(key=(-166, 8), weight=1.8421018192417273, enabled=True)
	DefaultConnectionGene(key=(-166, 11), weight=-1.0228610291591764, enabled=True)
	DefaultConnectionGene(key=(-165, 0), weight=-0.6370177782679716, enabled=True)
	DefaultConnectionGene(key=(-165, 1), weight=0.30632426131086465, enabled=False)
	DefaultConnectionGene(key=(-165, 2), weight=-0.19250612239459458, enabled=False)
	DefaultConnectionGene(key=(-165, 3), weight=-0.34949094894584604, enabled=False)
	DefaultConnectionGene(key=(-165, 5), weight=-1.2043896192120591, enabled=False)
	DefaultConnectionGene(key=(-165, 10), weight=2.322307835521181, enabled=False)
	DefaultConnectionGene(key=(-164, 1), weight=-1.776428068125756, enabled=False)
	DefaultConnectionGene(key=(-164, 6), weight=0.40589935115462994, enabled=True)
	DefaultConnectionGene(key=(-164, 7), weight=-1.5524146410999742, enabled=False)
	DefaultConnectionGene(key=(-164, 8), weight=-0.32385083433421386, enabled=False)
	DefaultConnectionGene(key=(-164, 10), weight=-1.8649785847300635, enabled=False)
	DefaultConnectionGene(key=(-164, 11), weight=-1.7953200453790414, enabled=True)
	DefaultConnectionGene(key=(-163, 0), weight=-1.5884271386777973, enabled=True)
	DefaultConnectionGene(key=(-163, 1), weight=0.1395048074304914, enabled=False)
	DefaultConnectionGene(key=(-163, 3), weight=-0.3781055477969826, enabled=False)
	DefaultConnectionGene(key=(-163, 4), weight=0.7446451210578858, enabled=True)
	DefaultConnectionGene(key=(-163, 5), weight=1.669218539453448, enabled=True)
	DefaultConnectionGene(key=(-163, 6), weight=3.589172842886615, enabled=False)
	DefaultConnectionGene(key=(-163, 7), weight=0.29167375534978246, enabled=True)
	DefaultConnectionGene(key=(-163, 8), weight=-0.29795542109174755, enabled=False)
	DefaultConnectionGene(key=(-163, 10), weight=-2.2744500468315016, enabled=False)
	DefaultConnectionGene(key=(-162, 0), weight=-0.7049687795116815, enabled=True)
	DefaultConnectionGene(key=(-162, 1), weight=-0.12149178705195443, enabled=False)
	DefaultConnectionGene(key=(-162, 2), weight=0.34606155535609084, enabled=True)
	DefaultConnectionGene(key=(-162, 8), weight=1.1211878134988766, enabled=True)
	DefaultConnectionGene(key=(-162, 9), weight=-1.0415542099748916, enabled=True)
	DefaultConnectionGene(key=(-162, 11), weight=1.6122104529891355, enabled=False)
	DefaultConnectionGene(key=(-161, 2), weight=-0.06409880438840465, enabled=True)
	DefaultConnectionGene(key=(-161, 3), weight=1.5776856653777491, enabled=True)
	DefaultConnectionGene(key=(-161, 5), weight=-2.5049536542377138, enabled=True)
	DefaultConnectionGene(key=(-161, 7), weight=-0.12274625569286607, enabled=True)
	DefaultConnectionGene(key=(-161, 8), weight=-0.03503807145908461, enabled=False)
	DefaultConnectionGene(key=(-161, 11), weight=4.153898412512444, enabled=False)
	DefaultConnectionGene(key=(-160, 0), weight=-0.11086821167378885, enabled=False)
	DefaultConnectionGene(key=(-160, 2), weight=-0.25467681325239844, enabled=True)
	DefaultConnectionGene(key=(-160, 3), weight=0.5382490391779615, enabled=False)
	DefaultConnectionGene(key=(-160, 4), weight=0.9976103624403141, enabled=False)
	DefaultConnectionGene(key=(-160, 5), weight=-0.36072525395545485, enabled=False)
	DefaultConnectionGene(key=(-160, 6), weight=0.5581783806116142, enabled=False)
	DefaultConnectionGene(key=(-160, 8), weight=0.5832782522553949, enabled=False)
	DefaultConnectionGene(key=(-160, 9), weight=0.08965045263372615, enabled=False)
	DefaultConnectionGene(key=(-160, 1352), weight=0.6855646156821058, enabled=False)
	DefaultConnectionGene(key=(-159, 1), weight=2.2007670710335097, enabled=True)
	DefaultConnectionGene(key=(-159, 8), weight=-0.9946574842955288, enabled=True)
	DefaultConnectionGene(key=(-159, 11), weight=1.4113831858508354, enabled=True)
	DefaultConnectionGene(key=(-158, 1), weight=-1.3409390746099874, enabled=False)
	DefaultConnectionGene(key=(-158, 2), weight=0.12917389238479393, enabled=True)
	DefaultConnectionGene(key=(-158, 3), weight=-0.8114725835735067, enabled=True)
	DefaultConnectionGene(key=(-158, 4), weight=0.11834817762536559, enabled=True)
	DefaultConnectionGene(key=(-158, 6), weight=2.3053060437525357, enabled=False)
	DefaultConnectionGene(key=(-158, 9), weight=-0.39847822402901645, enabled=True)
	DefaultConnectionGene(key=(-158, 10), weight=0.8571945537148212, enabled=True)
	DefaultConnectionGene(key=(-158, 11), weight=-1.2993920615514223, enabled=True)
	DefaultConnectionGene(key=(-157, 0), weight=2.565083980614008, enabled=False)
	DefaultConnectionGene(key=(-157, 1), weight=5.808141617682227, enabled=False)
	DefaultConnectionGene(key=(-157, 2), weight=0.17711941496907663, enabled=True)
	DefaultConnectionGene(key=(-157, 3), weight=1.920925940396356, enabled=True)
	DefaultConnectionGene(key=(-157, 5), weight=0.47579181230951245, enabled=True)
	DefaultConnectionGene(key=(-157, 6), weight=-1.175893717254725, enabled=False)
	DefaultConnectionGene(key=(-157, 7), weight=-1.0427459964116539, enabled=False)
	DefaultConnectionGene(key=(-157, 8), weight=-2.077131577976118, enabled=False)
	DefaultConnectionGene(key=(-157, 11), weight=1.4675233798674436, enabled=False)
	DefaultConnectionGene(key=(-156, 3), weight=0.13695897695334622, enabled=True)
	DefaultConnectionGene(key=(-156, 4), weight=-0.6074333680011538, enabled=False)
	DefaultConnectionGene(key=(-156, 5), weight=-0.47168070618778835, enabled=False)
	DefaultConnectionGene(key=(-156, 6), weight=-1.8206652318602006, enabled=False)
	DefaultConnectionGene(key=(-156, 7), weight=1.4351490616599627, enabled=False)
	DefaultConnectionGene(key=(-156, 8), weight=-0.9619932184839278, enabled=True)
	DefaultConnectionGene(key=(-156, 10), weight=1.596847569652359, enabled=True)
	DefaultConnectionGene(key=(-156, 11), weight=2.08803467702887, enabled=False)
	DefaultConnectionGene(key=(-155, 1), weight=3.7625400185303204, enabled=True)
	DefaultConnectionGene(key=(-155, 3), weight=-5.393765902820328, enabled=False)
	DefaultConnectionGene(key=(-155, 5), weight=-0.9318007662932295, enabled=True)
	DefaultConnectionGene(key=(-155, 8), weight=-0.5966611620417737, enabled=True)
	DefaultConnectionGene(key=(-155, 9), weight=0.15110549810498078, enabled=True)
	DefaultConnectionGene(key=(-155, 10), weight=1.5612192325703356, enabled=True)
	DefaultConnectionGene(key=(-155, 11), weight=-0.302693157994922, enabled=False)
	DefaultConnectionGene(key=(-154, 1), weight=-2.0812925392861783, enabled=False)
	DefaultConnectionGene(key=(-154, 3), weight=1.9240046648384865, enabled=False)
	DefaultConnectionGene(key=(-154, 4), weight=0.994231123485589, enabled=True)
	DefaultConnectionGene(key=(-154, 5), weight=-0.9017423522369986, enabled=True)
	DefaultConnectionGene(key=(-154, 7), weight=0.18087775450572613, enabled=True)
	DefaultConnectionGene(key=(-154, 10), weight=1.7317598607023148, enabled=False)
	DefaultConnectionGene(key=(-153, 4), weight=-0.0704397026974517, enabled=True)
	DefaultConnectionGene(key=(-153, 6), weight=-0.9708631920263531, enabled=False)
	DefaultConnectionGene(key=(-153, 8), weight=1.461870127911754, enabled=True)
	DefaultConnectionGene(key=(-153, 11), weight=0.32956299315942383, enabled=False)
	DefaultConnectionGene(key=(-152, 1), weight=0.1931386238523944, enabled=True)
	DefaultConnectionGene(key=(-152, 2), weight=0.9323839605373301, enabled=True)
	DefaultConnectionGene(key=(-152, 4), weight=1.2250074773520516, enabled=True)
	DefaultConnectionGene(key=(-152, 6), weight=-0.5987657389236825, enabled=True)
	DefaultConnectionGene(key=(-152, 7), weight=-2.6435243958241506, enabled=True)
	DefaultConnectionGene(key=(-152, 8), weight=-1.6067702940015394, enabled=True)
	DefaultConnectionGene(key=(-151, 0), weight=1.469774158540595, enabled=True)
	DefaultConnectionGene(key=(-151, 1), weight=-1.789391926286741, enabled=True)
	DefaultConnectionGene(key=(-151, 4), weight=-1.2748391028231278, enabled=False)
	DefaultConnectionGene(key=(-151, 6), weight=0.035037037454142106, enabled=True)
	DefaultConnectionGene(key=(-151, 8), weight=-0.8352803267377127, enabled=True)
	DefaultConnectionGene(key=(-151, 9), weight=2.1335479474997308, enabled=True)
	DefaultConnectionGene(key=(-151, 10), weight=-0.7180833688067907, enabled=False)
	DefaultConnectionGene(key=(-151, 11), weight=0.31192907564965977, enabled=True)
	DefaultConnectionGene(key=(-150, 3), weight=0.02564631391166078, enabled=False)
	DefaultConnectionGene(key=(-150, 7), weight=-1.7593759396835749, enabled=False)
	DefaultConnectionGene(key=(-150, 11), weight=-0.692501558644473, enabled=False)
	DefaultConnectionGene(key=(-149, 1), weight=-0.1296713409746543, enabled=True)
	DefaultConnectionGene(key=(-149, 5), weight=-0.289153226828537, enabled=False)
	DefaultConnectionGene(key=(-149, 11), weight=2.126823141461555, enabled=True)
	DefaultConnectionGene(key=(-148, 1), weight=1.4827260682011845, enabled=False)
	DefaultConnectionGene(key=(-148, 9), weight=1.9282721584961227, enabled=False)
	DefaultConnectionGene(key=(-148, 11), weight=0.5591572683224558, enabled=False)
	DefaultConnectionGene(key=(-147, 3), weight=1.8194686800307123, enabled=False)
	DefaultConnectionGene(key=(-147, 4), weight=-3.434443015291052, enabled=False)
	DefaultConnectionGene(key=(-147, 5), weight=1.4959975102383243, enabled=True)
	DefaultConnectionGene(key=(-147, 7), weight=-0.22289863443879776, enabled=False)
	DefaultConnectionGene(key=(-146, 1), weight=-4.12745973015828, enabled=True)
	DefaultConnectionGene(key=(-146, 2), weight=0.5875032999342549, enabled=True)
	DefaultConnectionGene(key=(-146, 6), weight=-0.868417305049889, enabled=True)
	DefaultConnectionGene(key=(-146, 9), weight=-0.7672062690147327, enabled=True)
	DefaultConnectionGene(key=(-146, 10), weight=-0.35872859685842307, enabled=True)
	DefaultConnectionGene(key=(-146, 11), weight=0.9225564741847799, enabled=False)
	DefaultConnectionGene(key=(-145, 0), weight=0.9277160803454878, enabled=False)
	DefaultConnectionGene(key=(-145, 1), weight=0.6404003821225501, enabled=True)
	DefaultConnectionGene(key=(-145, 5), weight=0.29991442819157704, enabled=False)
	DefaultConnectionGene(key=(-145, 6), weight=-1.012329293353606, enabled=True)
	DefaultConnectionGene(key=(-145, 7), weight=-0.9267586213715889, enabled=True)
	DefaultConnectionGene(key=(-145, 10), weight=-1.8271631606190948, enabled=True)
	DefaultConnectionGene(key=(-145, 11), weight=-1.136275727480792, enabled=False)
	DefaultConnectionGene(key=(-144, 0), weight=1.1163837668457708, enabled=False)
	DefaultConnectionGene(key=(-144, 1), weight=0.7548786435493449, enabled=True)
	DefaultConnectionGene(key=(-144, 2), weight=-4.33287367330423, enabled=True)
	DefaultConnectionGene(key=(-144, 3), weight=1.663513450985651, enabled=False)
	DefaultConnectionGene(key=(-144, 5), weight=-0.0634644563746525, enabled=True)
	DefaultConnectionGene(key=(-144, 8), weight=1.9002399892904387, enabled=True)
	DefaultConnectionGene(key=(-144, 11), weight=0.5776679312397961, enabled=True)
	DefaultConnectionGene(key=(-143, 0), weight=-2.514856351301955, enabled=False)
	DefaultConnectionGene(key=(-143, 1), weight=-2.974122159377178, enabled=False)
	DefaultConnectionGene(key=(-143, 2), weight=-1.410459659309138, enabled=True)
	DefaultConnectionGene(key=(-143, 3), weight=2.0451827287877826, enabled=True)
	DefaultConnectionGene(key=(-143, 4), weight=-0.2129090885677581, enabled=False)
	DefaultConnectionGene(key=(-143, 6), weight=-0.2586013568872539, enabled=True)
	DefaultConnectionGene(key=(-143, 8), weight=1.1549960077701773, enabled=True)
	DefaultConnectionGene(key=(-143, 11), weight=-3.939897450514467, enabled=True)
	DefaultConnectionGene(key=(-142, 0), weight=0.23241873497778487, enabled=True)
	DefaultConnectionGene(key=(-142, 3), weight=-0.20758019354984997, enabled=True)
	DefaultConnectionGene(key=(-142, 6), weight=-3.5233745039740767, enabled=False)
	DefaultConnectionGene(key=(-142, 1294), weight=1.0, enabled=False)
	DefaultConnectionGene(key=(-141, 0), weight=0.7158577263171122, enabled=False)
	DefaultConnectionGene(key=(-141, 1), weight=-1.7656402634128574, enabled=True)
	DefaultConnectionGene(key=(-141, 4), weight=-0.5568791233560004, enabled=False)
	DefaultConnectionGene(key=(-141, 6), weight=0.07277079489313237, enabled=False)
	DefaultConnectionGene(key=(-141, 11), weight=-1.3100135098543786, enabled=True)
	DefaultConnectionGene(key=(-140, 1), weight=-0.031292232331617986, enabled=True)
	DefaultConnectionGene(key=(-140, 2), weight=-1.3490601597871716, enabled=True)
	DefaultConnectionGene(key=(-140, 4), weight=-4.586294327225165, enabled=False)
	DefaultConnectionGene(key=(-140, 5), weight=0.21364447639454165, enabled=True)
	DefaultConnectionGene(key=(-140, 6), weight=0.6066769989308219, enabled=True)
	DefaultConnectionGene(key=(-140, 7), weight=-0.6226268454699957, enabled=True)
	DefaultConnectionGene(key=(-140, 8), weight=-1.0340877104572073, enabled=False)
	DefaultConnectionGene(key=(-140, 9), weight=0.32305680836695905, enabled=True)
	DefaultConnectionGene(key=(-140, 11), weight=-1.6455424091362516, enabled=False)
	DefaultConnectionGene(key=(-140, 879), weight=1.5971255636912178, enabled=False)
	DefaultConnectionGene(key=(-139, 4), weight=-1.0510632740574102, enabled=False)
	DefaultConnectionGene(key=(-139, 6), weight=-0.7254005665003672, enabled=False)
	DefaultConnectionGene(key=(-139, 7), weight=0.6181078875535547, enabled=True)
	DefaultConnectionGene(key=(-139, 9), weight=0.08077837817471262, enabled=False)
	DefaultConnectionGene(key=(-139, 10), weight=-0.71623854703819, enabled=False)
	DefaultConnectionGene(key=(-138, 0), weight=-0.5980261625677631, enabled=False)
	DefaultConnectionGene(key=(-138, 2), weight=-2.1190423015277093, enabled=False)
	DefaultConnectionGene(key=(-138, 8), weight=-2.5812534492004446, enabled=False)
	DefaultConnectionGene(key=(-138, 9), weight=1.2783010911334767, enabled=True)
	DefaultConnectionGene(key=(-138, 10), weight=-0.6113947738814195, enabled=False)
	DefaultConnectionGene(key=(-137, 5), weight=-2.279247374785623, enabled=True)
	DefaultConnectionGene(key=(-137, 6), weight=-0.40123859493084324, enabled=False)
	DefaultConnectionGene(key=(-137, 8), weight=-2.6456557458866876, enabled=True)
	DefaultConnectionGene(key=(-137, 9), weight=-0.21842365721380352, enabled=False)
	DefaultConnectionGene(key=(-137, 11), weight=1.3218561997273444, enabled=False)
	DefaultConnectionGene(key=(-136, 0), weight=2.351168439079441, enabled=False)
	DefaultConnectionGene(key=(-136, 3), weight=-0.0700894088263817, enabled=True)
	DefaultConnectionGene(key=(-136, 4), weight=2.352528145013472, enabled=True)
	DefaultConnectionGene(key=(-136, 6), weight=1.0130470932997342, enabled=True)
	DefaultConnectionGene(key=(-136, 10), weight=0.9674151794790605, enabled=False)
	DefaultConnectionGene(key=(-136, 11), weight=0.8288530955373452, enabled=False)
	DefaultConnectionGene(key=(-135, 2), weight=-0.2700397867729727, enabled=True)
	DefaultConnectionGene(key=(-135, 3), weight=-0.333295852344189, enabled=True)
	DefaultConnectionGene(key=(-135, 5), weight=0.17656268701490063, enabled=False)
	DefaultConnectionGene(key=(-135, 6), weight=2.4580746581786603, enabled=False)
	DefaultConnectionGene(key=(-135, 9), weight=-0.10825799696341457, enabled=True)
	DefaultConnectionGene(key=(-135, 11), weight=-1.565436819075403, enabled=False)
	DefaultConnectionGene(key=(-134, 0), weight=-1.1514708513732916, enabled=True)
	DefaultConnectionGene(key=(-134, 1), weight=0.058474270295718916, enabled=False)
	DefaultConnectionGene(key=(-134, 5), weight=0.7041038225221515, enabled=False)
	DefaultConnectionGene(key=(-134, 6), weight=-1.1379275331174994, enabled=True)
	DefaultConnectionGene(key=(-134, 7), weight=3.598327356765517, enabled=True)
	DefaultConnectionGene(key=(-134, 8), weight=-2.2932848092116376, enabled=True)
	DefaultConnectionGene(key=(-134, 10), weight=0.32392264264343357, enabled=False)
	DefaultConnectionGene(key=(-133, 3), weight=0.10243417000215983, enabled=True)
	DefaultConnectionGene(key=(-133, 7), weight=4.87038556273503, enabled=True)
	DefaultConnectionGene(key=(-133, 8), weight=-2.0064416395503533, enabled=False)
	DefaultConnectionGene(key=(-133, 10), weight=-1.0874097209415534, enabled=False)
	DefaultConnectionGene(key=(-132, 2), weight=2.2773344393075705, enabled=True)
	DefaultConnectionGene(key=(-132, 6), weight=-1.3766473321267543, enabled=True)
	DefaultConnectionGene(key=(-132, 9), weight=-2.3799840977949382, enabled=True)
	DefaultConnectionGene(key=(-131, 0), weight=-0.31837599973264125, enabled=False)
	DefaultConnectionGene(key=(-131, 3), weight=0.3473704060830271, enabled=False)
	DefaultConnectionGene(key=(-131, 6), weight=0.728278764649052, enabled=False)
	DefaultConnectionGene(key=(-131, 9), weight=0.09345709794164334, enabled=True)
	DefaultConnectionGene(key=(-131, 10), weight=-4.287359620211643, enabled=True)
	DefaultConnectionGene(key=(-130, 0), weight=0.9250216482551309, enabled=True)
	DefaultConnectionGene(key=(-130, 3), weight=0.4408450560842753, enabled=False)
	DefaultConnectionGene(key=(-130, 5), weight=0.5587807347045741, enabled=True)
	DefaultConnectionGene(key=(-130, 7), weight=-1.4107787952808677, enabled=False)
	DefaultConnectionGene(key=(-130, 8), weight=2.3398652727308256, enabled=False)
	DefaultConnectionGene(key=(-129, 1), weight=0.7984413545050097, enabled=False)
	DefaultConnectionGene(key=(-129, 3), weight=0.8479478115511166, enabled=False)
	DefaultConnectionGene(key=(-129, 4), weight=0.9855054954456808, enabled=True)
	DefaultConnectionGene(key=(-129, 6), weight=-1.5888474892834896, enabled=False)
	DefaultConnectionGene(key=(-129, 8), weight=1.092967375677971, enabled=False)
	DefaultConnectionGene(key=(-128, 0), weight=0.6805470729095978, enabled=True)
	DefaultConnectionGene(key=(-128, 1), weight=3.1658900942591237, enabled=False)
	DefaultConnectionGene(key=(-128, 2), weight=-3.2373863495604867, enabled=False)
	DefaultConnectionGene(key=(-128, 3), weight=0.3719976535384465, enabled=True)
	DefaultConnectionGene(key=(-128, 4), weight=1.4074023358547874, enabled=True)
	DefaultConnectionGene(key=(-128, 7), weight=-1.7620955730685217, enabled=False)
	DefaultConnectionGene(key=(-128, 260), weight=2.517717126233518, enabled=True)
	DefaultConnectionGene(key=(-127, 0), weight=0.7945346415514842, enabled=False)
	DefaultConnectionGene(key=(-127, 6), weight=3.180967722506844, enabled=False)
	DefaultConnectionGene(key=(-127, 8), weight=2.260925570779799, enabled=False)
	DefaultConnectionGene(key=(-126, 1), weight=1.189871826013171, enabled=True)
	DefaultConnectionGene(key=(-126, 2), weight=1.4066561227255145, enabled=True)
	DefaultConnectionGene(key=(-126, 3), weight=-0.7027944689231318, enabled=False)
	DefaultConnectionGene(key=(-126, 4), weight=-1.5202317366378537, enabled=True)
	DefaultConnectionGene(key=(-126, 7), weight=1.8661911089936156, enabled=True)
	DefaultConnectionGene(key=(-126, 10), weight=2.114420325899184, enabled=False)
	DefaultConnectionGene(key=(-126, 11), weight=0.6534406038700211, enabled=False)
	DefaultConnectionGene(key=(-125, 2), weight=0.537046939434216, enabled=True)
	DefaultConnectionGene(key=(-125, 3), weight=1.9876027039217274, enabled=False)
	DefaultConnectionGene(key=(-125, 5), weight=1.5702224819886663, enabled=False)
	DefaultConnectionGene(key=(-125, 6), weight=-1.2425453334058378, enabled=False)
	DefaultConnectionGene(key=(-125, 7), weight=0.5113356606794153, enabled=True)
	DefaultConnectionGene(key=(-125, 8), weight=1.0788442118139943, enabled=True)
	DefaultConnectionGene(key=(-125, 9), weight=0.7233933628844144, enabled=True)
	DefaultConnectionGene(key=(-124, 1), weight=0.5885403604855811, enabled=True)
	DefaultConnectionGene(key=(-124, 2), weight=3.116000081900779, enabled=True)
	DefaultConnectionGene(key=(-124, 5), weight=-1.606035735360626, enabled=False)
	DefaultConnectionGene(key=(-124, 9), weight=-0.7225898223959549, enabled=True)
	DefaultConnectionGene(key=(-124, 10), weight=-2.7391906359406013, enabled=False)
	DefaultConnectionGene(key=(-123, 3), weight=0.27778378399916026, enabled=True)
	DefaultConnectionGene(key=(-123, 4), weight=-1.239690787719126, enabled=True)
	DefaultConnectionGene(key=(-123, 5), weight=-0.41525305712735355, enabled=False)
	DefaultConnectionGene(key=(-123, 9), weight=-0.05211885354467616, enabled=False)
	DefaultConnectionGene(key=(-123, 11), weight=2.330617162299151, enabled=False)
	DefaultConnectionGene(key=(-122, 0), weight=-2.617717360915773, enabled=False)
	DefaultConnectionGene(key=(-122, 4), weight=-0.4924803057581695, enabled=False)
	DefaultConnectionGene(key=(-122, 6), weight=1.395697285615618, enabled=True)
	DefaultConnectionGene(key=(-122, 7), weight=-1.7738547116575247, enabled=False)
	DefaultConnectionGene(key=(-122, 8), weight=2.65751265435386, enabled=False)
	DefaultConnectionGene(key=(-122, 9), weight=-1.4491275161161248, enabled=True)
	DefaultConnectionGene(key=(-121, 0), weight=2.356991854827684, enabled=False)
	DefaultConnectionGene(key=(-121, 1), weight=-1.1417862098645213, enabled=False)
	DefaultConnectionGene(key=(-121, 2), weight=-0.021661470197264443, enabled=True)
	DefaultConnectionGene(key=(-121, 3), weight=-1.5397290386132527, enabled=True)
	DefaultConnectionGene(key=(-121, 4), weight=3.6514913587927937, enabled=True)
	DefaultConnectionGene(key=(-121, 6), weight=-0.9888939541197299, enabled=False)
	DefaultConnectionGene(key=(-120, 0), weight=0.5398598503122439, enabled=True)
	DefaultConnectionGene(key=(-120, 1), weight=-0.8215936686303846, enabled=True)
	DefaultConnectionGene(key=(-120, 3), weight=-2.4355112136173664, enabled=False)
	DefaultConnectionGene(key=(-120, 8), weight=1.1335264121553876, enabled=False)
	DefaultConnectionGene(key=(-119, 1), weight=0.7334649534615998, enabled=True)
	DefaultConnectionGene(key=(-119, 2), weight=-1.259153136848366, enabled=True)
	DefaultConnectionGene(key=(-119, 4), weight=0.6627377055263722, enabled=False)
	DefaultConnectionGene(key=(-119, 5), weight=4.132271892150915, enabled=True)
	DefaultConnectionGene(key=(-119, 6), weight=0.36842773405325757, enabled=True)
	DefaultConnectionGene(key=(-119, 7), weight=1.39257752552425, enabled=False)
	DefaultConnectionGene(key=(-119, 9), weight=1.1810947064516277, enabled=False)
	DefaultConnectionGene(key=(-118, 0), weight=0.7998367497691364, enabled=False)
	DefaultConnectionGene(key=(-118, 1), weight=2.9371061740581736, enabled=True)
	DefaultConnectionGene(key=(-118, 2), weight=-0.8696255449580307, enabled=False)
	DefaultConnectionGene(key=(-118, 3), weight=0.26632522020348337, enabled=False)
	DefaultConnectionGene(key=(-118, 5), weight=-1.1337984330484003, enabled=True)
	DefaultConnectionGene(key=(-118, 9), weight=-0.6146592084078959, enabled=False)
	DefaultConnectionGene(key=(-118, 11), weight=1.355114626136567, enabled=False)
	DefaultConnectionGene(key=(-117, 0), weight=-3.1093353491564413, enabled=True)
	DefaultConnectionGene(key=(-117, 2), weight=-0.9091048465906583, enabled=True)
	DefaultConnectionGene(key=(-117, 3), weight=1.1333803119063712, enabled=True)
	DefaultConnectionGene(key=(-117, 5), weight=-0.6618954148660975, enabled=True)
	DefaultConnectionGene(key=(-117, 6), weight=-0.8132409547057432, enabled=True)
	DefaultConnectionGene(key=(-117, 9), weight=1.7852802647325707, enabled=False)
	DefaultConnectionGene(key=(-117, 11), weight=-1.1978335303115206, enabled=True)
	DefaultConnectionGene(key=(-116, 0), weight=-2.079658607958238, enabled=True)
	DefaultConnectionGene(key=(-116, 1), weight=0.23863471896107308, enabled=False)
	DefaultConnectionGene(key=(-116, 4), weight=-1.5276849011972942, enabled=False)
	DefaultConnectionGene(key=(-116, 5), weight=-0.019046515820088365, enabled=False)
	DefaultConnectionGene(key=(-116, 7), weight=-0.7033736053840264, enabled=False)
	DefaultConnectionGene(key=(-116, 8), weight=-0.10759578019994237, enabled=False)
	DefaultConnectionGene(key=(-116, 9), weight=-1.7304478852740344, enabled=True)
	DefaultConnectionGene(key=(-116, 10), weight=-1.8888429088657848, enabled=True)
	DefaultConnectionGene(key=(-115, 0), weight=0.8914909133903179, enabled=False)
	DefaultConnectionGene(key=(-115, 2), weight=-0.09633535665259457, enabled=False)
	DefaultConnectionGene(key=(-115, 3), weight=-2.0753409312037485, enabled=True)
	DefaultConnectionGene(key=(-115, 4), weight=1.2880587068108265, enabled=True)
	DefaultConnectionGene(key=(-115, 6), weight=-0.18438111379248556, enabled=True)
	DefaultConnectionGene(key=(-115, 7), weight=-5.239177909485861, enabled=True)
	DefaultConnectionGene(key=(-115, 11), weight=-0.41157271716229893, enabled=True)
	DefaultConnectionGene(key=(-115, 30), weight=0.576386039837653, enabled=True)
	DefaultConnectionGene(key=(-114, 0), weight=1.8143782751101405, enabled=True)
	DefaultConnectionGene(key=(-114, 2), weight=0.12738998067393037, enabled=True)
	DefaultConnectionGene(key=(-114, 3), weight=-2.9261216609415377, enabled=False)
	DefaultConnectionGene(key=(-114, 6), weight=0.08643063876306291, enabled=True)
	DefaultConnectionGene(key=(-114, 7), weight=0.4916307709881281, enabled=True)
	DefaultConnectionGene(key=(-113, 0), weight=1.0224742203439356, enabled=False)
	DefaultConnectionGene(key=(-113, 1), weight=-0.040246058019513115, enabled=False)
	DefaultConnectionGene(key=(-113, 2), weight=1.88546156153448, enabled=True)
	DefaultConnectionGene(key=(-113, 8), weight=2.004741747262894, enabled=False)
	DefaultConnectionGene(key=(-113, 9), weight=-1.4456338071874903, enabled=True)
	DefaultConnectionGene(key=(-112, 1), weight=-0.5929941315480882, enabled=True)
	DefaultConnectionGene(key=(-112, 2), weight=0.15510983122826816, enabled=False)
	DefaultConnectionGene(key=(-112, 3), weight=0.7562411718926936, enabled=True)
	DefaultConnectionGene(key=(-112, 5), weight=-0.9102165956115424, enabled=True)
	DefaultConnectionGene(key=(-112, 9), weight=0.8599548036161149, enabled=True)
	DefaultConnectionGene(key=(-112, 11), weight=-0.6606489753730735, enabled=False)
	DefaultConnectionGene(key=(-111, 0), weight=0.011987383743404356, enabled=False)
	DefaultConnectionGene(key=(-111, 1), weight=0.29850512020250497, enabled=True)
	DefaultConnectionGene(key=(-111, 3), weight=-0.6408213698217966, enabled=False)
	DefaultConnectionGene(key=(-111, 6), weight=5.388523241182634, enabled=True)
	DefaultConnectionGene(key=(-111, 7), weight=-0.5589555378263953, enabled=True)
	DefaultConnectionGene(key=(-111, 11), weight=1.5290898403840973, enabled=True)
	DefaultConnectionGene(key=(-110, 0), weight=-1.7531457945912372, enabled=False)
	DefaultConnectionGene(key=(-110, 4), weight=-0.09292982391984791, enabled=False)
	DefaultConnectionGene(key=(-110, 8), weight=-1.5941537874386664, enabled=False)
	DefaultConnectionGene(key=(-109, 0), weight=0.8765705303740653, enabled=False)
	DefaultConnectionGene(key=(-109, 1), weight=-2.8971281457448708, enabled=True)
	DefaultConnectionGene(key=(-109, 3), weight=1.2206903327726053, enabled=False)
	DefaultConnectionGene(key=(-109, 8), weight=0.7792167261456212, enabled=True)
	DefaultConnectionGene(key=(-109, 9), weight=-0.5765124938699638, enabled=False)
	DefaultConnectionGene(key=(-109, 10), weight=-1.5849909992678124, enabled=False)
	DefaultConnectionGene(key=(-108, 1), weight=-0.5322143144798602, enabled=False)
	DefaultConnectionGene(key=(-108, 2), weight=-0.70795195162712, enabled=False)
	DefaultConnectionGene(key=(-108, 4), weight=-1.758323141978089, enabled=False)
	DefaultConnectionGene(key=(-108, 7), weight=-2.0891460381006612, enabled=True)
	DefaultConnectionGene(key=(-108, 8), weight=1.4307364863153142, enabled=True)
	DefaultConnectionGene(key=(-108, 11), weight=1.99699556948928, enabled=True)
	DefaultConnectionGene(key=(-107, 5), weight=-0.25930357916432356, enabled=False)
	DefaultConnectionGene(key=(-107, 6), weight=0.6334692708350445, enabled=True)
	DefaultConnectionGene(key=(-107, 7), weight=2.217613146419367, enabled=True)
	DefaultConnectionGene(key=(-107, 11), weight=3.4783478602321534, enabled=True)
	DefaultConnectionGene(key=(-106, 1), weight=0.8953791312051627, enabled=False)
	DefaultConnectionGene(key=(-106, 2), weight=1.0026572906506508, enabled=True)
	DefaultConnectionGene(key=(-106, 4), weight=1.792667899516025, enabled=True)
	DefaultConnectionGene(key=(-106, 5), weight=-0.0031302632263580515, enabled=True)
	DefaultConnectionGene(key=(-106, 6), weight=-0.40094987363275836, enabled=False)
	DefaultConnectionGene(key=(-106, 7), weight=-2.473316202897913, enabled=True)
	DefaultConnectionGene(key=(-106, 8), weight=-1.3252896703316146, enabled=False)
	DefaultConnectionGene(key=(-106, 11), weight=0.22627471498740218, enabled=False)
	DefaultConnectionGene(key=(-105, 1), weight=1.3285298815163413, enabled=False)
	DefaultConnectionGene(key=(-105, 2), weight=0.6733921182282139, enabled=True)
	DefaultConnectionGene(key=(-105, 3), weight=-0.3733104984618706, enabled=True)
	DefaultConnectionGene(key=(-105, 4), weight=2.6482704837733113, enabled=False)
	DefaultConnectionGene(key=(-105, 8), weight=4.050005568498423, enabled=True)
	DefaultConnectionGene(key=(-105, 9), weight=-1.634693586022282, enabled=False)
	DefaultConnectionGene(key=(-104, 4), weight=0.193289641097414, enabled=True)
	DefaultConnectionGene(key=(-104, 5), weight=0.8045431876018478, enabled=True)
	DefaultConnectionGene(key=(-104, 8), weight=-1.4595463735187335, enabled=True)
	DefaultConnectionGene(key=(-104, 10), weight=0.49865449397030825, enabled=False)
	DefaultConnectionGene(key=(-104, 11), weight=-0.11266213195790153, enabled=True)
	DefaultConnectionGene(key=(-103, 0), weight=-1.8922119092936576, enabled=False)
	DefaultConnectionGene(key=(-103, 3), weight=0.06804222354335142, enabled=True)
	DefaultConnectionGene(key=(-103, 5), weight=1.2365538712649975, enabled=True)
	DefaultConnectionGene(key=(-103, 6), weight=0.012671016899841181, enabled=False)
	DefaultConnectionGene(key=(-103, 9), weight=-1.1354942247615971, enabled=False)
	DefaultConnectionGene(key=(-102, 5), weight=1.9942029514367106, enabled=False)
	DefaultConnectionGene(key=(-102, 7), weight=-2.1629093135644193, enabled=True)
	DefaultConnectionGene(key=(-101, 0), weight=1.6716588258484726, enabled=False)
	DefaultConnectionGene(key=(-101, 3), weight=-1.9270457387423643, enabled=True)
	DefaultConnectionGene(key=(-101, 5), weight=-0.8095450575233274, enabled=False)
	DefaultConnectionGene(key=(-101, 6), weight=3.076187457949855, enabled=False)
	DefaultConnectionGene(key=(-101, 7), weight=-0.6083110429779628, enabled=False)
	DefaultConnectionGene(key=(-101, 8), weight=0.8090334445774884, enabled=True)
	DefaultConnectionGene(key=(-101, 9), weight=0.749090192220225, enabled=False)
	DefaultConnectionGene(key=(-101, 11), weight=-0.5938617585641986, enabled=False)
	DefaultConnectionGene(key=(-100, 0), weight=-0.9633264752301014, enabled=True)
	DefaultConnectionGene(key=(-100, 1), weight=-0.4041103273775909, enabled=True)
	DefaultConnectionGene(key=(-100, 3), weight=0.34913642262682965, enabled=False)
	DefaultConnectionGene(key=(-100, 5), weight=0.7891159327363152, enabled=True)
	DefaultConnectionGene(key=(-100, 7), weight=2.008990374891698, enabled=False)
	DefaultConnectionGene(key=(-100, 9), weight=-0.3155106157994323, enabled=True)
	DefaultConnectionGene(key=(-100, 10), weight=-2.207350439656083, enabled=True)
	DefaultConnectionGene(key=(-99, 0), weight=-0.6091506633724157, enabled=True)
	DefaultConnectionGene(key=(-99, 1), weight=-1.9925566481222945, enabled=False)
	DefaultConnectionGene(key=(-99, 2), weight=4.142051263478887, enabled=True)
	DefaultConnectionGene(key=(-99, 3), weight=-0.531231326251289, enabled=True)
	DefaultConnectionGene(key=(-99, 9), weight=0.050881424880234394, enabled=True)
	DefaultConnectionGene(key=(-99, 10), weight=0.24035070981921436, enabled=True)
	DefaultConnectionGene(key=(-99, 11), weight=-0.3774928313212508, enabled=True)
	DefaultConnectionGene(key=(-98, 2), weight=1.071693312594967, enabled=True)
	DefaultConnectionGene(key=(-98, 4), weight=-1.205804606671496, enabled=True)
	DefaultConnectionGene(key=(-98, 5), weight=2.1779961198951243, enabled=False)
	DefaultConnectionGene(key=(-98, 6), weight=-3.6791196574225453, enabled=True)
	DefaultConnectionGene(key=(-98, 8), weight=1.760370581221618, enabled=False)
	DefaultConnectionGene(key=(-98, 9), weight=1.2239668638519337, enabled=False)
	DefaultConnectionGene(key=(-98, 10), weight=-2.0346362060545644, enabled=True)
	DefaultConnectionGene(key=(-97, 1), weight=-1.2462846225618476, enabled=False)
	DefaultConnectionGene(key=(-97, 2), weight=-0.7178891314255369, enabled=True)
	DefaultConnectionGene(key=(-97, 3), weight=-1.7688099447649999, enabled=False)
	DefaultConnectionGene(key=(-97, 4), weight=-1.4917500301008806, enabled=True)
	DefaultConnectionGene(key=(-97, 5), weight=0.8833296005805064, enabled=False)
	DefaultConnectionGene(key=(-97, 8), weight=1.152138811961552, enabled=True)
	DefaultConnectionGene(key=(-97, 9), weight=-0.7545666228497546, enabled=True)
	DefaultConnectionGene(key=(-96, 0), weight=-0.0670496606667983, enabled=True)
	DefaultConnectionGene(key=(-96, 2), weight=1.0112173074849742, enabled=True)
	DefaultConnectionGene(key=(-96, 3), weight=0.43085359782211263, enabled=False)
	DefaultConnectionGene(key=(-96, 6), weight=0.46024087490563426, enabled=False)
	DefaultConnectionGene(key=(-96, 8), weight=0.18111910394052316, enabled=True)
	DefaultConnectionGene(key=(-96, 9), weight=2.171998672967153, enabled=True)
	DefaultConnectionGene(key=(-96, 10), weight=1.3300056919227716, enabled=False)
	DefaultConnectionGene(key=(-96, 11), weight=1.9254642850249406, enabled=False)
	DefaultConnectionGene(key=(-95, 3), weight=-0.46593048434523504, enabled=True)
	DefaultConnectionGene(key=(-95, 10), weight=0.26081051498958696, enabled=False)
	DefaultConnectionGene(key=(-95, 11), weight=1.9752484315050605, enabled=True)
	DefaultConnectionGene(key=(-94, 0), weight=0.31769459022154195, enabled=False)
	DefaultConnectionGene(key=(-94, 4), weight=-0.6603426361454994, enabled=False)
	DefaultConnectionGene(key=(-94, 6), weight=0.3726620372383738, enabled=False)
	DefaultConnectionGene(key=(-94, 7), weight=-0.6246450081448602, enabled=False)
	DefaultConnectionGene(key=(-94, 8), weight=-0.9598848614295898, enabled=False)
	DefaultConnectionGene(key=(-94, 9), weight=1.0686297609256634, enabled=False)
	DefaultConnectionGene(key=(-94, 10), weight=1.2720857017225269, enabled=True)
	DefaultConnectionGene(key=(-93, 0), weight=-1.2705968505350516, enabled=False)
	DefaultConnectionGene(key=(-93, 4), weight=0.9643721148693203, enabled=False)
	DefaultConnectionGene(key=(-93, 6), weight=2.354280707899358, enabled=True)
	DefaultConnectionGene(key=(-93, 8), weight=1.180994277341515, enabled=False)
	DefaultConnectionGene(key=(-93, 10), weight=-3.5631123572516588, enabled=False)
	DefaultConnectionGene(key=(-92, 3), weight=-1.064353202092045, enabled=True)
	DefaultConnectionGene(key=(-92, 6), weight=-1.8458496443112642, enabled=False)
	DefaultConnectionGene(key=(-91, 1), weight=1.453012599792656, enabled=False)
	DefaultConnectionGene(key=(-91, 2), weight=-0.1562329256273271, enabled=False)
	DefaultConnectionGene(key=(-91, 5), weight=1.7963245041997284, enabled=True)
	DefaultConnectionGene(key=(-91, 6), weight=2.1207498866104375, enabled=False)
	DefaultConnectionGene(key=(-91, 8), weight=-1.7388705937815727, enabled=True)
	DefaultConnectionGene(key=(-91, 10), weight=0.1655858919701504, enabled=True)
	DefaultConnectionGene(key=(-90, 1), weight=-1.7722001281329205, enabled=False)
	DefaultConnectionGene(key=(-90, 2), weight=0.8657468100882381, enabled=False)
	DefaultConnectionGene(key=(-90, 4), weight=-0.05795959230126624, enabled=False)
	DefaultConnectionGene(key=(-90, 6), weight=-0.4349658364954392, enabled=True)
	DefaultConnectionGene(key=(-90, 10), weight=-1.730801550511942, enabled=True)
	DefaultConnectionGene(key=(-89, 0), weight=0.41543364943531824, enabled=True)
	DefaultConnectionGene(key=(-89, 2), weight=1.623403887305986, enabled=True)
	DefaultConnectionGene(key=(-89, 3), weight=-0.46843515917718026, enabled=False)
	DefaultConnectionGene(key=(-89, 4), weight=-1.1892136723216689, enabled=False)
	DefaultConnectionGene(key=(-89, 5), weight=1.2566329370593887, enabled=True)
	DefaultConnectionGene(key=(-89, 7), weight=1.3602379691497708, enabled=False)
	DefaultConnectionGene(key=(-89, 8), weight=-0.8919649181889968, enabled=False)
	DefaultConnectionGene(key=(-89, 9), weight=0.4072821341919416, enabled=False)
	DefaultConnectionGene(key=(-89, 10), weight=-3.103596391325887, enabled=False)
	DefaultConnectionGene(key=(-88, 2), weight=1.0479208544128453, enabled=True)
	DefaultConnectionGene(key=(-88, 5), weight=1.4173519495710616, enabled=False)
	DefaultConnectionGene(key=(-88, 6), weight=1.9296311902972703, enabled=False)
	DefaultConnectionGene(key=(-88, 7), weight=2.2345515067145922, enabled=False)
	DefaultConnectionGene(key=(-88, 8), weight=0.8176509843753268, enabled=False)
	DefaultConnectionGene(key=(-88, 9), weight=-0.7309766638585445, enabled=False)
	DefaultConnectionGene(key=(-87, 0), weight=-4.001250823241688, enabled=False)
	DefaultConnectionGene(key=(-87, 1), weight=-1.6355878898918168, enabled=True)
	DefaultConnectionGene(key=(-87, 10), weight=-0.13021518500380197, enabled=False)
	DefaultConnectionGene(key=(-87, 11), weight=2.627386760951587, enabled=True)
	DefaultConnectionGene(key=(-86, 0), weight=2.3842750731902584, enabled=True)
	DefaultConnectionGene(key=(-86, 1), weight=-1.7380574630480694, enabled=False)
	DefaultConnectionGene(key=(-86, 2), weight=-3.4405325849245325, enabled=False)
	DefaultConnectionGene(key=(-86, 4), weight=-0.20145229495794179, enabled=True)
	DefaultConnectionGene(key=(-86, 5), weight=0.2757327029709147, enabled=False)
	DefaultConnectionGene(key=(-86, 6), weight=-0.6868076774967519, enabled=False)
	DefaultConnectionGene(key=(-86, 9), weight=-0.6040822454902858, enabled=False)
	DefaultConnectionGene(key=(-86, 11), weight=0.7152777023277836, enabled=False)
	DefaultConnectionGene(key=(-85, 0), weight=-1.7479375923153484, enabled=True)
	DefaultConnectionGene(key=(-85, 1), weight=0.6063411610608317, enabled=True)
	DefaultConnectionGene(key=(-85, 4), weight=0.016856878124968, enabled=True)
	DefaultConnectionGene(key=(-85, 5), weight=-2.440716928930422, enabled=False)
	DefaultConnectionGene(key=(-85, 6), weight=3.5290996486198387, enabled=False)
	DefaultConnectionGene(key=(-85, 8), weight=1.8720369987334196, enabled=False)
	DefaultConnectionGene(key=(-85, 9), weight=1.0827307403416215, enabled=True)
	DefaultConnectionGene(key=(-85, 10), weight=-0.805845895986347, enabled=False)
	DefaultConnectionGene(key=(-84, 4), weight=0.32181149152959215, enabled=False)
	DefaultConnectionGene(key=(-84, 8), weight=-2.7770514688895793, enabled=False)
	DefaultConnectionGene(key=(-84, 11), weight=1.097553238843492, enabled=True)
	DefaultConnectionGene(key=(-83, 1), weight=2.7397713528846435, enabled=False)
	DefaultConnectionGene(key=(-83, 2), weight=-0.011557742707027141, enabled=False)
	DefaultConnectionGene(key=(-83, 4), weight=2.9562281887097184, enabled=False)
	DefaultConnectionGene(key=(-83, 9), weight=2.1743664050486227, enabled=True)
	DefaultConnectionGene(key=(-83, 10), weight=1.0960592504149547, enabled=False)
	DefaultConnectionGene(key=(-82, 1), weight=-0.08309799734674739, enabled=True)
	DefaultConnectionGene(key=(-82, 2), weight=-0.33508885482642653, enabled=True)
	DefaultConnectionGene(key=(-82, 4), weight=-0.9862186582782847, enabled=False)
	DefaultConnectionGene(key=(-82, 5), weight=2.586384774227871, enabled=False)
	DefaultConnectionGene(key=(-82, 6), weight=0.4672593026091475, enabled=False)
	DefaultConnectionGene(key=(-82, 8), weight=2.596959534714449, enabled=True)
	DefaultConnectionGene(key=(-82, 9), weight=1.0461148467485226, enabled=False)
	DefaultConnectionGene(key=(-82, 10), weight=0.1702834308991072, enabled=False)
	DefaultConnectionGene(key=(-82, 11), weight=-0.16103677021398216, enabled=True)
	DefaultConnectionGene(key=(-81, 0), weight=3.5002241170041937, enabled=False)
	DefaultConnectionGene(key=(-81, 2), weight=0.03940382933404374, enabled=True)
	DefaultConnectionGene(key=(-81, 3), weight=0.799965111486953, enabled=False)
	DefaultConnectionGene(key=(-81, 6), weight=0.8825061168779469, enabled=True)
	DefaultConnectionGene(key=(-81, 7), weight=-2.093873547216112, enabled=True)
	DefaultConnectionGene(key=(-81, 8), weight=-0.20003675620917016, enabled=True)
	DefaultConnectionGene(key=(-81, 9), weight=2.84609701354092, enabled=False)
	DefaultConnectionGene(key=(-80, 0), weight=0.9846368249215908, enabled=False)
	DefaultConnectionGene(key=(-80, 3), weight=1.2905602570941208, enabled=False)
	DefaultConnectionGene(key=(-80, 7), weight=-0.6883326512316204, enabled=False)
	DefaultConnectionGene(key=(-79, 1), weight=-2.562395753056978, enabled=True)
	DefaultConnectionGene(key=(-79, 2), weight=-2.0998757230268694, enabled=True)
	DefaultConnectionGene(key=(-79, 3), weight=-0.21807261713354015, enabled=False)
	DefaultConnectionGene(key=(-79, 4), weight=-2.5049810172228932, enabled=True)
	DefaultConnectionGene(key=(-79, 5), weight=-0.6127642242514397, enabled=False)
	DefaultConnectionGene(key=(-79, 6), weight=-0.9559616484345054, enabled=False)
	DefaultConnectionGene(key=(-79, 8), weight=0.12251050536467803, enabled=True)
	DefaultConnectionGene(key=(-79, 11), weight=-2.9243751119700003, enabled=True)
	DefaultConnectionGene(key=(-78, 0), weight=-0.16042347249706923, enabled=True)
	DefaultConnectionGene(key=(-78, 2), weight=0.7185622717694529, enabled=True)
	DefaultConnectionGene(key=(-78, 3), weight=2.7948216430496764, enabled=False)
	DefaultConnectionGene(key=(-78, 4), weight=-1.3192459568253385, enabled=False)
	DefaultConnectionGene(key=(-78, 5), weight=1.008747659913886, enabled=False)
	DefaultConnectionGene(key=(-78, 6), weight=-2.431353676265732, enabled=False)
	DefaultConnectionGene(key=(-78, 7), weight=-1.139102678996651, enabled=True)
	DefaultConnectionGene(key=(-78, 9), weight=-0.6739226605267684, enabled=True)
	DefaultConnectionGene(key=(-78, 11), weight=-1.5116991226903562, enabled=True)
	DefaultConnectionGene(key=(-77, 0), weight=-2.606397974718, enabled=False)
	DefaultConnectionGene(key=(-77, 1), weight=-1.7006868782416504, enabled=True)
	DefaultConnectionGene(key=(-77, 4), weight=-0.7438106772133602, enabled=True)
	DefaultConnectionGene(key=(-77, 6), weight=-0.18504649565605102, enabled=False)
	DefaultConnectionGene(key=(-77, 7), weight=0.905865478791643, enabled=True)
	DefaultConnectionGene(key=(-77, 11), weight=-0.8543165807395581, enabled=True)
	DefaultConnectionGene(key=(-76, 4), weight=-1.140655738028889, enabled=False)
	DefaultConnectionGene(key=(-76, 7), weight=1.2846049509599784, enabled=True)
	DefaultConnectionGene(key=(-76, 8), weight=1.0294796908642003, enabled=False)
	DefaultConnectionGene(key=(-76, 10), weight=4.994175380324314, enabled=True)
	DefaultConnectionGene(key=(-76, 11), weight=-0.6940273410137536, enabled=True)
	DefaultConnectionGene(key=(-75, 0), weight=-1.1873941881395202, enabled=False)
	DefaultConnectionGene(key=(-75, 1), weight=-3.0857409923740815, enabled=False)
	DefaultConnectionGene(key=(-75, 3), weight=-3.4062718706073913, enabled=True)
	DefaultConnectionGene(key=(-75, 4), weight=-0.5848939895284047, enabled=False)
	DefaultConnectionGene(key=(-75, 7), weight=-2.0900676224928834, enabled=False)
	DefaultConnectionGene(key=(-75, 8), weight=-1.0139355307405409, enabled=False)
	DefaultConnectionGene(key=(-75, 10), weight=0.736915908142137, enabled=False)
	DefaultConnectionGene(key=(-75, 11), weight=-1.3827171777899987, enabled=True)
	DefaultConnectionGene(key=(-74, 1), weight=0.6276249541486472, enabled=True)
	DefaultConnectionGene(key=(-74, 4), weight=-0.5111832341355881, enabled=True)
	DefaultConnectionGene(key=(-74, 5), weight=2.4271756906468727, enabled=True)
	DefaultConnectionGene(key=(-74, 6), weight=-0.9154746189796454, enabled=False)
	DefaultConnectionGene(key=(-74, 7), weight=-0.972177012278552, enabled=False)
	DefaultConnectionGene(key=(-74, 8), weight=1.075797204480656, enabled=True)
	DefaultConnectionGene(key=(-74, 10), weight=-1.4737568734513042, enabled=True)
	DefaultConnectionGene(key=(-73, 0), weight=-0.4031604590947444, enabled=True)
	DefaultConnectionGene(key=(-73, 1), weight=0.005538149637861867, enabled=True)
	DefaultConnectionGene(key=(-73, 2), weight=2.7260454700807295, enabled=True)
	DefaultConnectionGene(key=(-73, 3), weight=-0.46297048817399133, enabled=True)
	DefaultConnectionGene(key=(-73, 7), weight=-0.40348015976991425, enabled=False)
	DefaultConnectionGene(key=(-73, 11), weight=1.1601399231496345, enabled=False)
	DefaultConnectionGene(key=(-72, 2), weight=-2.2971420625790095, enabled=False)
	DefaultConnectionGene(key=(-72, 3), weight=4.08538537843524, enabled=False)
	DefaultConnectionGene(key=(-72, 4), weight=-0.0015695153101953907, enabled=False)
	DefaultConnectionGene(key=(-72, 6), weight=0.1166180045794532, enabled=True)
	DefaultConnectionGene(key=(-72, 8), weight=-0.5092964987003216, enabled=False)
	DefaultConnectionGene(key=(-72, 10), weight=0.32039669811160354, enabled=False)
	DefaultConnectionGene(key=(-71, 0), weight=-0.6655515835012017, enabled=True)
	DefaultConnectionGene(key=(-71, 1), weight=0.2604909743994224, enabled=False)
	DefaultConnectionGene(key=(-71, 2), weight=-2.2315875888820766, enabled=True)
	DefaultConnectionGene(key=(-71, 5), weight=0.9505145204900297, enabled=True)
	DefaultConnectionGene(key=(-71, 8), weight=1.7668955199998206, enabled=False)
	DefaultConnectionGene(key=(-71, 11), weight=-2.6607463034011776, enabled=True)
	DefaultConnectionGene(key=(-71, 1402), weight=1.0, enabled=True)
	DefaultConnectionGene(key=(-70, 7), weight=0.3948350066686294, enabled=False)
	DefaultConnectionGene(key=(-70, 8), weight=-0.39969658942784436, enabled=True)
	DefaultConnectionGene(key=(-69, 1), weight=2.453715380414686, enabled=False)
	DefaultConnectionGene(key=(-69, 4), weight=-1.262383591570312, enabled=True)
	DefaultConnectionGene(key=(-69, 8), weight=-0.4951058375917674, enabled=False)
	DefaultConnectionGene(key=(-68, 1), weight=0.2849168479452632, enabled=True)
	DefaultConnectionGene(key=(-68, 3), weight=0.5746373408440517, enabled=False)
	DefaultConnectionGene(key=(-67, 0), weight=-0.9369009047216222, enabled=True)
	DefaultConnectionGene(key=(-67, 9), weight=-0.3074061361659497, enabled=True)
	DefaultConnectionGene(key=(-66, 4), weight=-0.14861876283665987, enabled=False)
	DefaultConnectionGene(key=(-66, 5), weight=0.09154883431955627, enabled=False)
	DefaultConnectionGene(key=(-66, 6), weight=0.554293752913091, enabled=True)
	DefaultConnectionGene(key=(-66, 7), weight=-1.756007715899344, enabled=True)
	DefaultConnectionGene(key=(-66, 10), weight=0.6060329003001985, enabled=False)
	DefaultConnectionGene(key=(-66, 11), weight=2.2401181644393082, enabled=True)
	DefaultConnectionGene(key=(-65, 1), weight=1.2286617123604873, enabled=False)
	DefaultConnectionGene(key=(-65, 2), weight=0.20483701351638706, enabled=True)
	DefaultConnectionGene(key=(-65, 4), weight=1.0485377426272746, enabled=True)
	DefaultConnectionGene(key=(-65, 5), weight=-0.7807073545756584, enabled=False)
	DefaultConnectionGene(key=(-65, 7), weight=-1.375624624488873, enabled=False)
	DefaultConnectionGene(key=(-65, 9), weight=-2.613802081185755, enabled=True)
	DefaultConnectionGene(key=(-64, 0), weight=0.02084139038598745, enabled=False)
	DefaultConnectionGene(key=(-64, 1), weight=2.3001810449853908, enabled=True)
	DefaultConnectionGene(key=(-64, 2), weight=1.0625696317786737, enabled=True)
	DefaultConnectionGene(key=(-64, 4), weight=-2.228370057062712, enabled=True)
	DefaultConnectionGene(key=(-64, 8), weight=-2.665054106346556, enabled=False)
	DefaultConnectionGene(key=(-64, 10), weight=-2.72039147630043, enabled=False)
	DefaultConnectionGene(key=(-64, 11), weight=-1.0207986301154865, enabled=True)
	DefaultConnectionGene(key=(-63, 1), weight=-0.45879763163662024, enabled=True)
	DefaultConnectionGene(key=(-63, 3), weight=1.3429787112914675, enabled=False)
	DefaultConnectionGene(key=(-63, 4), weight=1.4546156059272495, enabled=False)
	DefaultConnectionGene(key=(-63, 5), weight=5.901177736289215, enabled=True)
	DefaultConnectionGene(key=(-63, 7), weight=-1.2562254532816184, enabled=True)
	DefaultConnectionGene(key=(-63, 10), weight=1.524232027717923, enabled=True)
	DefaultConnectionGene(key=(-62, 0), weight=2.5983157856314336, enabled=True)
	DefaultConnectionGene(key=(-62, 1), weight=-0.4108446085669671, enabled=False)
	DefaultConnectionGene(key=(-62, 2), weight=-0.2829938834326036, enabled=True)
	DefaultConnectionGene(key=(-62, 4), weight=0.6215900887771523, enabled=True)
	DefaultConnectionGene(key=(-62, 5), weight=-3.974805214819079, enabled=True)
	DefaultConnectionGene(key=(-62, 6), weight=-0.5179515356094366, enabled=False)
	DefaultConnectionGene(key=(-62, 7), weight=1.1782702507159697, enabled=True)
	DefaultConnectionGene(key=(-61, 0), weight=1.9214094446224337, enabled=False)
	DefaultConnectionGene(key=(-61, 2), weight=1.5494860023765258, enabled=True)
	DefaultConnectionGene(key=(-61, 5), weight=1.5916986337706844, enabled=False)
	DefaultConnectionGene(key=(-61, 8), weight=-3.5733432140826134, enabled=False)
	DefaultConnectionGene(key=(-61, 9), weight=1.752807659964969, enabled=True)
	DefaultConnectionGene(key=(-61, 11), weight=0.0037792056504585836, enabled=True)
	DefaultConnectionGene(key=(-61, 260), weight=-0.42197531595929394, enabled=False)
	DefaultConnectionGene(key=(-60, 3), weight=-1.7582026416147127, enabled=False)
	DefaultConnectionGene(key=(-60, 5), weight=-1.3774085523401869, enabled=False)
	DefaultConnectionGene(key=(-60, 6), weight=-0.17620313411880084, enabled=True)
	DefaultConnectionGene(key=(-60, 7), weight=-1.6754236212781297, enabled=True)
	DefaultConnectionGene(key=(-60, 8), weight=0.14796140575687905, enabled=False)
	DefaultConnectionGene(key=(-60, 9), weight=-1.1188340969370625, enabled=False)
	DefaultConnectionGene(key=(-59, 1), weight=1.6442444548535762, enabled=True)
	DefaultConnectionGene(key=(-59, 3), weight=-1.5086633443084252, enabled=True)
	DefaultConnectionGene(key=(-59, 6), weight=-1.176875387969788, enabled=False)
	DefaultConnectionGene(key=(-59, 10), weight=-1.2027017681246417, enabled=True)
	DefaultConnectionGene(key=(-58, 1), weight=0.3655202210535189, enabled=False)
	DefaultConnectionGene(key=(-58, 3), weight=1.1692726601386418, enabled=True)
	DefaultConnectionGene(key=(-58, 6), weight=-0.6008714596993536, enabled=True)
	DefaultConnectionGene(key=(-58, 9), weight=1.4587776654071205, enabled=True)
	DefaultConnectionGene(key=(-58, 10), weight=-0.10524173801137107, enabled=True)
	DefaultConnectionGene(key=(-57, 0), weight=-2.8597478239613023, enabled=True)
	DefaultConnectionGene(key=(-57, 1), weight=-0.5158948596806537, enabled=False)
	DefaultConnectionGene(key=(-57, 4), weight=0.3084888651714778, enabled=False)
	DefaultConnectionGene(key=(-56, 0), weight=3.2592675812602625, enabled=False)
	DefaultConnectionGene(key=(-56, 4), weight=2.7644703711482435, enabled=False)
	DefaultConnectionGene(key=(-56, 6), weight=0.7948500353316983, enabled=False)
	DefaultConnectionGene(key=(-56, 7), weight=-0.9428901464718418, enabled=False)
	DefaultConnectionGene(key=(-56, 8), weight=-1.680153220821139, enabled=True)
	DefaultConnectionGene(key=(-56, 9), weight=1.9216802985078936, enabled=False)
	DefaultConnectionGene(key=(-55, 3), weight=-0.7261927629708608, enabled=True)
	DefaultConnectionGene(key=(-55, 4), weight=1.116927705819851, enabled=True)
	DefaultConnectionGene(key=(-55, 6), weight=1.7367232815771618, enabled=False)
	DefaultConnectionGene(key=(-55, 7), weight=-3.3420616365670255, enabled=False)
	DefaultConnectionGene(key=(-55, 8), weight=2.09873856492323, enabled=False)
	DefaultConnectionGene(key=(-55, 9), weight=2.0803191278825173, enabled=True)
	DefaultConnectionGene(key=(-55, 10), weight=-1.480960435612491, enabled=False)
	DefaultConnectionGene(key=(-55, 11), weight=0.5374879894923812, enabled=False)
	DefaultConnectionGene(key=(-54, 0), weight=-0.11984600807166279, enabled=False)
	DefaultConnectionGene(key=(-54, 2), weight=-0.8217938810449547, enabled=False)
	DefaultConnectionGene(key=(-54, 3), weight=-2.9614435803270656, enabled=True)
	DefaultConnectionGene(key=(-54, 5), weight=0.20817874337746578, enabled=False)
	DefaultConnectionGene(key=(-54, 7), weight=-1.0582283781993982, enabled=False)
	DefaultConnectionGene(key=(-54, 8), weight=0.3412610643342172, enabled=False)
	DefaultConnectionGene(key=(-54, 9), weight=2.14015724470624, enabled=True)
	DefaultConnectionGene(key=(-54, 11), weight=1.0116812568592197, enabled=True)
	DefaultConnectionGene(key=(-54, 209), weight=-0.19363356464430176, enabled=False)
	DefaultConnectionGene(key=(-53, 0), weight=2.183175059737327, enabled=False)
	DefaultConnectionGene(key=(-53, 1), weight=1.2944445329141658, enabled=True)
	DefaultConnectionGene(key=(-53, 3), weight=0.39247413496953987, enabled=True)
	DefaultConnectionGene(key=(-53, 5), weight=-0.3313184073571803, enabled=False)
	DefaultConnectionGene(key=(-53, 6), weight=1.4273398143390577, enabled=True)
	DefaultConnectionGene(key=(-53, 9), weight=0.2293850697255667, enabled=False)
	DefaultConnectionGene(key=(-53, 11), weight=0.062358111932481454, enabled=False)
	DefaultConnectionGene(key=(-52, 0), weight=1.259838713818369, enabled=False)
	DefaultConnectionGene(key=(-52, 1), weight=1.7077463882760657, enabled=False)
	DefaultConnectionGene(key=(-52, 2), weight=0.8178628797904552, enabled=False)
	DefaultConnectionGene(key=(-52, 3), weight=0.28406293048356746, enabled=False)
	DefaultConnectionGene(key=(-52, 4), weight=-2.0976158669696003, enabled=True)
	DefaultConnectionGene(key=(-52, 8), weight=-4.971714798860392, enabled=True)
	DefaultConnectionGene(key=(-51, 2), weight=1.6190566877091144, enabled=True)
	DefaultConnectionGene(key=(-51, 3), weight=-0.7614696861672545, enabled=True)
	DefaultConnectionGene(key=(-51, 5), weight=2.0146554848948144, enabled=True)
	DefaultConnectionGene(key=(-51, 8), weight=-0.9223442635274115, enabled=True)
	DefaultConnectionGene(key=(-51, 9), weight=-0.25631817731061535, enabled=True)
	DefaultConnectionGene(key=(-51, 10), weight=0.6810161999481835, enabled=False)
	DefaultConnectionGene(key=(-51, 11), weight=0.3885280893695129, enabled=False)
	DefaultConnectionGene(key=(-50, 0), weight=-0.576844506873282, enabled=True)
	DefaultConnectionGene(key=(-50, 2), weight=1.6401445439070548, enabled=True)
	DefaultConnectionGene(key=(-50, 3), weight=2.048618963551474, enabled=False)
	DefaultConnectionGene(key=(-50, 5), weight=-1.40135509863126, enabled=False)
	DefaultConnectionGene(key=(-50, 6), weight=0.12090668245130531, enabled=True)
	DefaultConnectionGene(key=(-50, 8), weight=2.025237490563022, enabled=False)
	DefaultConnectionGene(key=(-50, 9), weight=1.8516954849228182, enabled=False)
	DefaultConnectionGene(key=(-50, 11), weight=-0.6345368156153608, enabled=False)
	DefaultConnectionGene(key=(-49, 0), weight=1.7174289850432416, enabled=False)
	DefaultConnectionGene(key=(-49, 3), weight=-0.5572943352206409, enabled=True)
	DefaultConnectionGene(key=(-49, 5), weight=0.9552545580159568, enabled=True)
	DefaultConnectionGene(key=(-49, 6), weight=0.6696833885071569, enabled=True)
	DefaultConnectionGene(key=(-48, 2), weight=-0.6632002903049805, enabled=False)
	DefaultConnectionGene(key=(-48, 3), weight=1.2969672672122565, enabled=False)
	DefaultConnectionGene(key=(-48, 4), weight=0.1617551970183382, enabled=False)
	DefaultConnectionGene(key=(-48, 10), weight=0.1823372637624776, enabled=True)
	DefaultConnectionGene(key=(-48, 11), weight=-1.0073061520017008, enabled=True)
	DefaultConnectionGene(key=(-47, 0), weight=0.07401004344525253, enabled=False)
	DefaultConnectionGene(key=(-47, 2), weight=-0.9108716702705048, enabled=False)
	DefaultConnectionGene(key=(-47, 7), weight=-1.4691645777444402, enabled=True)
	DefaultConnectionGene(key=(-47, 9), weight=2.6996662093736195, enabled=True)
	DefaultConnectionGene(key=(-47, 11), weight=-0.987427983810206, enabled=False)
	DefaultConnectionGene(key=(-46, 0), weight=0.5735942308553391, enabled=True)
	DefaultConnectionGene(key=(-46, 1), weight=-0.935788106496052, enabled=True)
	DefaultConnectionGene(key=(-46, 2), weight=1.9006450040259648, enabled=False)
	DefaultConnectionGene(key=(-46, 3), weight=-1.214753659828961, enabled=True)
	DefaultConnectionGene(key=(-46, 4), weight=-0.03068335350508722, enabled=False)
	DefaultConnectionGene(key=(-46, 5), weight=2.3146478289180576, enabled=False)
	DefaultConnectionGene(key=(-46, 6), weight=-0.8804253882592468, enabled=True)
	DefaultConnectionGene(key=(-46, 7), weight=1.1884012877210228, enabled=True)
	DefaultConnectionGene(key=(-46, 11), weight=0.6126303395576941, enabled=False)
	DefaultConnectionGene(key=(-45, 2), weight=2.67044226558364, enabled=True)
	DefaultConnectionGene(key=(-45, 4), weight=-1.0374769894865818, enabled=False)
	DefaultConnectionGene(key=(-45, 6), weight=-0.18634377765021298, enabled=False)
	DefaultConnectionGene(key=(-45, 7), weight=-1.5833196221018266, enabled=True)
	DefaultConnectionGene(key=(-45, 9), weight=-2.142114331698142, enabled=True)
	DefaultConnectionGene(key=(-45, 11), weight=-1.488267646871199, enabled=True)
	DefaultConnectionGene(key=(-44, 0), weight=-1.5943420956588237, enabled=False)
	DefaultConnectionGene(key=(-44, 4), weight=2.0220673957925084, enabled=False)
	DefaultConnectionGene(key=(-44, 5), weight=0.12369101703768742, enabled=False)
	DefaultConnectionGene(key=(-44, 6), weight=-1.3535568752737408, enabled=True)
	DefaultConnectionGene(key=(-44, 7), weight=-2.27416538051056, enabled=False)
	DefaultConnectionGene(key=(-44, 9), weight=-1.2960105381803932, enabled=False)
	DefaultConnectionGene(key=(-44, 11), weight=1.5918665510744796, enabled=True)
	DefaultConnectionGene(key=(-43, 0), weight=1.478593168416198, enabled=False)
	DefaultConnectionGene(key=(-43, 1), weight=-0.04921923081385432, enabled=False)
	DefaultConnectionGene(key=(-43, 3), weight=-1.4087376257732926, enabled=False)
	DefaultConnectionGene(key=(-43, 4), weight=0.9274470416644736, enabled=False)
	DefaultConnectionGene(key=(-43, 5), weight=-1.09107489565409, enabled=False)
	DefaultConnectionGene(key=(-43, 6), weight=-0.9430365694292399, enabled=False)
	DefaultConnectionGene(key=(-43, 7), weight=-0.2830243075748134, enabled=True)
	DefaultConnectionGene(key=(-43, 10), weight=1.1990346757074388, enabled=True)
	DefaultConnectionGene(key=(-42, 1), weight=2.559978038519644, enabled=False)
	DefaultConnectionGene(key=(-42, 2), weight=4.476168468126655, enabled=True)
	DefaultConnectionGene(key=(-42, 4), weight=-1.8239984493151467, enabled=True)
	DefaultConnectionGene(key=(-42, 5), weight=2.199795125851055, enabled=True)
	DefaultConnectionGene(key=(-42, 9), weight=0.22268256027302175, enabled=False)
	DefaultConnectionGene(key=(-42, 10), weight=-0.7581291743419729, enabled=False)
	DefaultConnectionGene(key=(-41, 2), weight=1.052771182620283, enabled=False)
	DefaultConnectionGene(key=(-41, 5), weight=-1.6554339420054673, enabled=False)
	DefaultConnectionGene(key=(-41, 7), weight=-3.810836006535601, enabled=False)
	DefaultConnectionGene(key=(-41, 8), weight=0.7171333883820277, enabled=False)
	DefaultConnectionGene(key=(-41, 10), weight=-0.5851597553826213, enabled=False)
	DefaultConnectionGene(key=(-41, 11), weight=0.6958985621076414, enabled=False)
	DefaultConnectionGene(key=(-40, 0), weight=0.5167338968028596, enabled=False)
	DefaultConnectionGene(key=(-40, 1), weight=0.5852519386526092, enabled=False)
	DefaultConnectionGene(key=(-40, 2), weight=-0.8414364761774118, enabled=False)
	DefaultConnectionGene(key=(-40, 3), weight=1.5759213505497205, enabled=False)
	DefaultConnectionGene(key=(-40, 5), weight=0.4263608094002903, enabled=True)
	DefaultConnectionGene(key=(-40, 6), weight=-0.7065496034064911, enabled=False)
	DefaultConnectionGene(key=(-40, 9), weight=1.2360436099991865, enabled=False)
	DefaultConnectionGene(key=(-40, 10), weight=0.5121549984968744, enabled=True)
	DefaultConnectionGene(key=(-39, 0), weight=0.14564483135872097, enabled=True)
	DefaultConnectionGene(key=(-39, 7), weight=1.239326724254554, enabled=False)
	DefaultConnectionGene(key=(-39, 10), weight=4.45863969212906, enabled=False)
	DefaultConnectionGene(key=(-39, 11), weight=0.10529265164878021, enabled=True)
	DefaultConnectionGene(key=(-38, 0), weight=-1.3467894016926745, enabled=False)
	DefaultConnectionGene(key=(-38, 1), weight=0.37322622662393645, enabled=True)
	DefaultConnectionGene(key=(-38, 2), weight=1.885182862126469, enabled=True)
	DefaultConnectionGene(key=(-38, 5), weight=-0.7345312771292369, enabled=False)
	DefaultConnectionGene(key=(-38, 6), weight=-1.1103541244332047, enabled=False)
	DefaultConnectionGene(key=(-38, 7), weight=-0.3385274962578907, enabled=False)
	DefaultConnectionGene(key=(-38, 10), weight=1.346142995785044, enabled=True)
	DefaultConnectionGene(key=(-38, 11), weight=-0.2307960634846767, enabled=False)
	DefaultConnectionGene(key=(-37, 2), weight=-1.2528486088524886, enabled=False)
	DefaultConnectionGene(key=(-37, 4), weight=0.611681568669642, enabled=False)
	DefaultConnectionGene(key=(-37, 7), weight=2.4321424306444435, enabled=False)
	DefaultConnectionGene(key=(-37, 9), weight=0.01093365582889834, enabled=True)
	DefaultConnectionGene(key=(-37, 10), weight=3.4970453009211817, enabled=True)
	DefaultConnectionGene(key=(-37, 11), weight=-1.0103749786876717, enabled=True)
	DefaultConnectionGene(key=(-36, 2), weight=1.5355390506063984, enabled=False)
	DefaultConnectionGene(key=(-36, 3), weight=0.3701921085063403, enabled=False)
	DefaultConnectionGene(key=(-36, 6), weight=3.752155019278434, enabled=True)
	DefaultConnectionGene(key=(-36, 10), weight=0.9295905348047562, enabled=False)
	DefaultConnectionGene(key=(-35, 0), weight=-0.8806983042193028, enabled=True)
	DefaultConnectionGene(key=(-35, 1), weight=-0.31773408512418844, enabled=False)
	DefaultConnectionGene(key=(-35, 2), weight=-1.8504672182174247, enabled=True)
	DefaultConnectionGene(key=(-35, 4), weight=0.11258501026554057, enabled=True)
	DefaultConnectionGene(key=(-35, 5), weight=-0.22321012977279434, enabled=False)
	DefaultConnectionGene(key=(-35, 7), weight=2.196882264817331, enabled=True)
	DefaultConnectionGene(key=(-35, 8), weight=-0.2482088893157405, enabled=True)
	DefaultConnectionGene(key=(-35, 9), weight=0.9393836476348985, enabled=True)
	DefaultConnectionGene(key=(-35, 11), weight=-1.2545697641813989, enabled=True)
	DefaultConnectionGene(key=(-34, 2), weight=-1.322542864658696, enabled=True)
	DefaultConnectionGene(key=(-34, 3), weight=0.10662734767674192, enabled=False)
	DefaultConnectionGene(key=(-34, 4), weight=-0.7395869075322995, enabled=False)
	DefaultConnectionGene(key=(-34, 5), weight=2.203828850952299, enabled=False)
	DefaultConnectionGene(key=(-34, 6), weight=-0.5204997815666439, enabled=True)
	DefaultConnectionGene(key=(-34, 7), weight=0.7516474397940875, enabled=True)
	DefaultConnectionGene(key=(-34, 8), weight=-0.13511322811593995, enabled=False)
	DefaultConnectionGene(key=(-34, 9), weight=-2.1670494332597476, enabled=False)
	DefaultConnectionGene(key=(-34, 10), weight=-0.37147517567911925, enabled=True)
	DefaultConnectionGene(key=(-34, 11), weight=-2.399773240489986, enabled=False)
	DefaultConnectionGene(key=(-33, 0), weight=-1.357193400484185, enabled=False)
	DefaultConnectionGene(key=(-33, 2), weight=-0.3925077383113531, enabled=True)
	DefaultConnectionGene(key=(-33, 3), weight=0.08861958938090256, enabled=True)
	DefaultConnectionGene(key=(-33, 5), weight=1.7301797417938545, enabled=False)
	DefaultConnectionGene(key=(-33, 7), weight=-2.5319163192707204, enabled=False)
	DefaultConnectionGene(key=(-33, 8), weight=2.596499928231468, enabled=False)
	DefaultConnectionGene(key=(-33, 10), weight=3.9648210459531343, enabled=True)
	DefaultConnectionGene(key=(-33, 11), weight=-0.8500950155736459, enabled=True)
	DefaultConnectionGene(key=(-32, 0), weight=0.9098121391888876, enabled=False)
	DefaultConnectionGene(key=(-32, 1), weight=-0.5188392051590337, enabled=True)
	DefaultConnectionGene(key=(-32, 2), weight=2.1097360104380796, enabled=True)
	DefaultConnectionGene(key=(-32, 5), weight=-0.14806823138991435, enabled=True)
	DefaultConnectionGene(key=(-32, 6), weight=0.750623407623273, enabled=True)
	DefaultConnectionGene(key=(-32, 9), weight=2.192458993862937, enabled=True)
	DefaultConnectionGene(key=(-32, 11), weight=-1.3593870328298026, enabled=False)
	DefaultConnectionGene(key=(-31, 0), weight=0.06538526602893786, enabled=True)
	DefaultConnectionGene(key=(-31, 2), weight=-2.2333369745197933, enabled=True)
	DefaultConnectionGene(key=(-31, 5), weight=-3.4527908636876874, enabled=False)
	DefaultConnectionGene(key=(-31, 6), weight=2.797790815845801, enabled=True)
	DefaultConnectionGene(key=(-31, 8), weight=0.0157835587656035, enabled=False)
	DefaultConnectionGene(key=(-31, 9), weight=-2.8611295692508762, enabled=True)
	DefaultConnectionGene(key=(-31, 10), weight=-1.9865553462070311, enabled=False)
	DefaultConnectionGene(key=(-30, 0), weight=1.0354563776066619, enabled=False)
	DefaultConnectionGene(key=(-30, 2), weight=-3.2334923183701147, enabled=True)
	DefaultConnectionGene(key=(-30, 4), weight=-0.1559397124588744, enabled=False)
	DefaultConnectionGene(key=(-30, 5), weight=-0.8382982979159335, enabled=True)
	DefaultConnectionGene(key=(-30, 6), weight=-3.6335567318418764, enabled=True)
	DefaultConnectionGene(key=(-30, 8), weight=1.094391508538275, enabled=True)
	DefaultConnectionGene(key=(-29, 0), weight=3.0132460971703607, enabled=False)
	DefaultConnectionGene(key=(-29, 2), weight=3.6562366893119345, enabled=False)
	DefaultConnectionGene(key=(-29, 3), weight=-2.9346245228798646, enabled=True)
	DefaultConnectionGene(key=(-29, 4), weight=-0.12362762757615756, enabled=False)
	DefaultConnectionGene(key=(-29, 5), weight=1.843708639958649, enabled=False)
	DefaultConnectionGene(key=(-29, 6), weight=-0.13224484005350604, enabled=False)
	DefaultConnectionGene(key=(-29, 9), weight=-3.190982182188835, enabled=True)
	DefaultConnectionGene(key=(-28, 3), weight=-0.7830474175427801, enabled=False)
	DefaultConnectionGene(key=(-28, 4), weight=-1.2322172929469917, enabled=True)
	DefaultConnectionGene(key=(-28, 7), weight=-0.474676312236907, enabled=True)
	DefaultConnectionGene(key=(-28, 9), weight=0.7639012736190699, enabled=False)
	DefaultConnectionGene(key=(-28, 10), weight=-1.0115921153102296, enabled=True)
	DefaultConnectionGene(key=(-28, 11), weight=2.3356226323468574, enabled=False)
	DefaultConnectionGene(key=(-28, 30), weight=-1.6909017595504914, enabled=False)
	DefaultConnectionGene(key=(-28, 652), weight=2.5134875422464003, enabled=False)
	DefaultConnectionGene(key=(-27, 0), weight=0.5774276892327908, enabled=False)
	DefaultConnectionGene(key=(-27, 1), weight=0.3063940868842857, enabled=False)
	DefaultConnectionGene(key=(-27, 2), weight=0.5958981021741169, enabled=True)
	DefaultConnectionGene(key=(-27, 3), weight=1.5990212243953554, enabled=True)
	DefaultConnectionGene(key=(-27, 4), weight=1.333276892484702, enabled=True)
	DefaultConnectionGene(key=(-27, 5), weight=1.0260170539743558, enabled=False)
	DefaultConnectionGene(key=(-27, 6), weight=0.18798006814826695, enabled=False)
	DefaultConnectionGene(key=(-27, 8), weight=-1.730614845855487, enabled=False)
	DefaultConnectionGene(key=(-27, 10), weight=3.2685829937460857, enabled=False)
	DefaultConnectionGene(key=(-26, 1), weight=2.6733390902134273, enabled=False)
	DefaultConnectionGene(key=(-26, 2), weight=-0.4791547790669838, enabled=False)
	DefaultConnectionGene(key=(-26, 5), weight=-0.5790789061392759, enabled=True)
	DefaultConnectionGene(key=(-26, 6), weight=0.7730845299608063, enabled=False)
	DefaultConnectionGene(key=(-26, 7), weight=0.7249082767299241, enabled=True)
	DefaultConnectionGene(key=(-26, 9), weight=-0.6958234049012706, enabled=True)
	DefaultConnectionGene(key=(-26, 10), weight=-0.4669231008297669, enabled=False)
	DefaultConnectionGene(key=(-26, 11), weight=0.7347533067191301, enabled=False)
	DefaultConnectionGene(key=(-25, 0), weight=0.102135489902492, enabled=False)
	DefaultConnectionGene(key=(-25, 2), weight=-0.17376780074157913, enabled=False)
	DefaultConnectionGene(key=(-25, 4), weight=0.4113152738436686, enabled=False)
	DefaultConnectionGene(key=(-25, 5), weight=0.7114304421778144, enabled=False)
	DefaultConnectionGene(key=(-25, 6), weight=-3.4670697892139084, enabled=False)
	DefaultConnectionGene(key=(-25, 9), weight=-1.719253989037771, enabled=False)
	DefaultConnectionGene(key=(-25, 10), weight=-0.6310672577736509, enabled=True)
	DefaultConnectionGene(key=(-24, 0), weight=0.4173059353765002, enabled=False)
	DefaultConnectionGene(key=(-24, 2), weight=-0.7749134500815049, enabled=True)
	DefaultConnectionGene(key=(-24, 7), weight=-0.3646292184172246, enabled=False)
	DefaultConnectionGene(key=(-24, 9), weight=-4.661292888514702, enabled=True)
	DefaultConnectionGene(key=(-24, 10), weight=-1.8750518815090338, enabled=False)
	DefaultConnectionGene(key=(-23, 1), weight=-0.3263349885490292, enabled=False)
	DefaultConnectionGene(key=(-23, 2), weight=-1.776072924601803, enabled=True)
	DefaultConnectionGene(key=(-23, 4), weight=1.999765424986235, enabled=False)
	DefaultConnectionGene(key=(-23, 6), weight=3.021568381266304, enabled=False)
	DefaultConnectionGene(key=(-23, 11), weight=2.0764304017687953, enabled=True)
	DefaultConnectionGene(key=(-22, 4), weight=3.2283952880180906, enabled=False)
	DefaultConnectionGene(key=(-22, 7), weight=-1.780695637491634, enabled=False)
	DefaultConnectionGene(key=(-22, 8), weight=-0.2719482317655011, enabled=False)
	DefaultConnectionGene(key=(-22, 9), weight=1.3589705011402455, enabled=False)
	DefaultConnectionGene(key=(-21, 4), weight=-2.7157952004186914, enabled=True)
	DefaultConnectionGene(key=(-21, 5), weight=-0.19059157240200947, enabled=False)
	DefaultConnectionGene(key=(-21, 6), weight=3.0247537439884247, enabled=True)
	DefaultConnectionGene(key=(-21, 8), weight=-3.2264604605723877, enabled=False)
	DefaultConnectionGene(key=(-21, 10), weight=-1.2599568702138269, enabled=True)
	DefaultConnectionGene(key=(-20, 0), weight=1.294021011452888, enabled=False)
	DefaultConnectionGene(key=(-20, 1), weight=1.7317808022149008, enabled=False)
	DefaultConnectionGene(key=(-20, 3), weight=0.3030416646824743, enabled=True)
	DefaultConnectionGene(key=(-20, 4), weight=-0.0860537309494199, enabled=True)
	DefaultConnectionGene(key=(-20, 6), weight=1.6918895730891563, enabled=True)
	DefaultConnectionGene(key=(-20, 8), weight=1.0389898622664935, enabled=False)
	DefaultConnectionGene(key=(-19, 0), weight=-0.485282398885162, enabled=False)
	DefaultConnectionGene(key=(-19, 1), weight=-0.42081519237432885, enabled=True)
	DefaultConnectionGene(key=(-19, 2), weight=-0.036009066068158124, enabled=False)
	DefaultConnectionGene(key=(-19, 3), weight=1.654035532864728, enabled=True)
	DefaultConnectionGene(key=(-19, 4), weight=-0.7486719367170768, enabled=True)
	DefaultConnectionGene(key=(-19, 5), weight=1.3969993552645616, enabled=False)
	DefaultConnectionGene(key=(-19, 6), weight=0.5405712589339554, enabled=True)
	DefaultConnectionGene(key=(-19, 10), weight=0.9377123720300599, enabled=False)
	DefaultConnectionGene(key=(-19, 11), weight=-0.12644056030682838, enabled=False)
	DefaultConnectionGene(key=(-18, 0), weight=1.1969504958209443, enabled=True)
	DefaultConnectionGene(key=(-18, 1), weight=1.5742529922310642, enabled=False)
	DefaultConnectionGene(key=(-18, 3), weight=-0.1499376680686187, enabled=False)
	DefaultConnectionGene(key=(-18, 4), weight=1.316751372941326, enabled=False)
	DefaultConnectionGene(key=(-18, 9), weight=-0.4716959757008374, enabled=False)
	DefaultConnectionGene(key=(-18, 10), weight=0.46837296467026795, enabled=False)
	DefaultConnectionGene(key=(-17, 0), weight=1.2938550587091164, enabled=True)
	DefaultConnectionGene(key=(-17, 1), weight=-1.0823730873444024, enabled=True)
	DefaultConnectionGene(key=(-17, 2), weight=0.007681907595403006, enabled=True)
	DefaultConnectionGene(key=(-17, 3), weight=-0.8250902546075761, enabled=True)
	DefaultConnectionGene(key=(-17, 5), weight=0.1115311819348147, enabled=False)
	DefaultConnectionGene(key=(-17, 6), weight=-1.9523985351288298, enabled=False)
	DefaultConnectionGene(key=(-17, 7), weight=1.1085140170623473, enabled=False)
	DefaultConnectionGene(key=(-17, 9), weight=0.027999963985993137, enabled=False)
	DefaultConnectionGene(key=(-17, 10), weight=0.0845162326116158, enabled=False)
	DefaultConnectionGene(key=(-17, 11), weight=-0.570022649796581, enabled=False)
	DefaultConnectionGene(key=(-16, 0), weight=-1.3901383433798165, enabled=False)
	DefaultConnectionGene(key=(-16, 2), weight=-1.3270002551898226, enabled=True)
	DefaultConnectionGene(key=(-16, 4), weight=1.670651043411191, enabled=False)
	DefaultConnectionGene(key=(-16, 6), weight=-0.599306985111995, enabled=False)
	DefaultConnectionGene(key=(-16, 9), weight=1.9684293999667193, enabled=False)
	DefaultConnectionGene(key=(-16, 10), weight=1.8225997016176352, enabled=False)
	DefaultConnectionGene(key=(-15, 1), weight=-1.1629138803898937, enabled=False)
	DefaultConnectionGene(key=(-15, 3), weight=0.4369473082538845, enabled=False)
	DefaultConnectionGene(key=(-15, 4), weight=0.8057414327331063, enabled=False)
	DefaultConnectionGene(key=(-15, 5), weight=0.8878225053101478, enabled=False)
	DefaultConnectionGene(key=(-15, 6), weight=0.5560068548954205, enabled=True)
	DefaultConnectionGene(key=(-15, 8), weight=-0.24719242300984512, enabled=True)
	DefaultConnectionGene(key=(-15, 10), weight=0.4060438084950033, enabled=True)
	DefaultConnectionGene(key=(-14, 0), weight=1.5184905630498184, enabled=False)
	DefaultConnectionGene(key=(-14, 3), weight=2.895000197462354, enabled=False)
	DefaultConnectionGene(key=(-14, 4), weight=-2.5904823948101474, enabled=True)
	DefaultConnectionGene(key=(-14, 5), weight=-0.5759081780773233, enabled=False)
	DefaultConnectionGene(key=(-13, 1), weight=2.004622387415707, enabled=False)
	DefaultConnectionGene(key=(-13, 4), weight=0.401841140136991, enabled=False)
	DefaultConnectionGene(key=(-13, 5), weight=0.45315160364303864, enabled=True)
	DefaultConnectionGene(key=(-13, 7), weight=0.8862209541848498, enabled=False)
	DefaultConnectionGene(key=(-13, 11), weight=-0.09812930144060283, enabled=True)
	DefaultConnectionGene(key=(-12, 0), weight=0.7537928760687171, enabled=False)
	DefaultConnectionGene(key=(-12, 1), weight=-1.6877563130511468, enabled=False)
	DefaultConnectionGene(key=(-12, 2), weight=-0.8885827210759638, enabled=True)
	DefaultConnectionGene(key=(-12, 3), weight=-2.2489592504713434, enabled=True)
	DefaultConnectionGene(key=(-12, 7), weight=0.8195914681308061, enabled=True)
	DefaultConnectionGene(key=(-12, 8), weight=3.8455131049332194, enabled=False)
	DefaultConnectionGene(key=(-12, 9), weight=-1.181414387232377, enabled=False)
	DefaultConnectionGene(key=(-11, 0), weight=0.22295315027806112, enabled=True)
	DefaultConnectionGene(key=(-11, 6), weight=-0.846882141984634, enabled=True)
	DefaultConnectionGene(key=(-11, 7), weight=0.5092856061392015, enabled=False)
	DefaultConnectionGene(key=(-11, 8), weight=1.5363468353790344, enabled=False)
	DefaultConnectionGene(key=(-11, 11), weight=-0.44569457390345635, enabled=True)
	DefaultConnectionGene(key=(-10, 0), weight=1.5334586834511967, enabled=False)
	DefaultConnectionGene(key=(-10, 1), weight=-1.1712222624162476, enabled=False)
	DefaultConnectionGene(key=(-10, 3), weight=-0.08104175405493397, enabled=False)
	DefaultConnectionGene(key=(-10, 4), weight=-3.0358443389064957, enabled=False)
	DefaultConnectionGene(key=(-10, 5), weight=-2.381060237390346, enabled=False)
	DefaultConnectionGene(key=(-10, 6), weight=-0.8646880395576935, enabled=True)
	DefaultConnectionGene(key=(-10, 7), weight=-4.320530930719374, enabled=True)
	DefaultConnectionGene(key=(-9, 0), weight=1.0757925956990861, enabled=False)
	DefaultConnectionGene(key=(-9, 1), weight=0.38974936451959397, enabled=False)
	DefaultConnectionGene(key=(-9, 3), weight=4.091558325408726, enabled=False)
	DefaultConnectionGene(key=(-9, 5), weight=3.1553537259042734, enabled=True)
	DefaultConnectionGene(key=(-9, 7), weight=3.4069866411666525, enabled=False)
	DefaultConnectionGene(key=(-9, 8), weight=-0.6787956416468806, enabled=True)
	DefaultConnectionGene(key=(-9, 9), weight=2.1416846183159355, enabled=True)
	DefaultConnectionGene(key=(-8, 1), weight=-0.42301588369879034, enabled=True)
	DefaultConnectionGene(key=(-8, 7), weight=1.428089768476749, enabled=False)
	DefaultConnectionGene(key=(-8, 8), weight=-0.174974632130309, enabled=True)
	DefaultConnectionGene(key=(-8, 9), weight=-1.565853667851203, enabled=False)
	DefaultConnectionGene(key=(-8, 10), weight=0.04801836422671879, enabled=False)
	DefaultConnectionGene(key=(-8, 11), weight=1.1160550768029696, enabled=True)
	DefaultConnectionGene(key=(-7, 2), weight=-0.42391998008436627, enabled=True)
	DefaultConnectionGene(key=(-7, 3), weight=-0.23509988108592225, enabled=True)
	DefaultConnectionGene(key=(-7, 6), weight=0.8053514464776942, enabled=False)
	DefaultConnectionGene(key=(-7, 7), weight=-0.6067754883165484, enabled=False)
	DefaultConnectionGene(key=(-7, 8), weight=-1.6294510921938221, enabled=True)
	DefaultConnectionGene(key=(-7, 9), weight=0.5520730145807864, enabled=False)
	DefaultConnectionGene(key=(-7, 10), weight=-0.06744036710448931, enabled=True)
	DefaultConnectionGene(key=(-7, 11), weight=-1.8436566544618285, enabled=False)
	DefaultConnectionGene(key=(-6, 0), weight=0.08982465251840596, enabled=False)
	DefaultConnectionGene(key=(-6, 1), weight=2.898071476558431, enabled=True)
	DefaultConnectionGene(key=(-6, 2), weight=-2.002925462152169, enabled=True)
	DefaultConnectionGene(key=(-6, 6), weight=0.5599373622780506, enabled=True)
	DefaultConnectionGene(key=(-6, 7), weight=0.17088028445683778, enabled=False)
	DefaultConnectionGene(key=(-6, 8), weight=-0.36320200504041045, enabled=False)
	DefaultConnectionGene(key=(-6, 9), weight=0.6865302207426798, enabled=False)
	DefaultConnectionGene(key=(-5, 2), weight=2.116724548376525, enabled=True)
	DefaultConnectionGene(key=(-5, 3), weight=1.4561966266356594, enabled=False)
	DefaultConnectionGene(key=(-5, 4), weight=-1.392277161787434, enabled=True)
	DefaultConnectionGene(key=(-5, 5), weight=-1.6580690176461463, enabled=False)
	DefaultConnectionGene(key=(-5, 6), weight=-1.9427481628517662, enabled=False)
	DefaultConnectionGene(key=(-5, 7), weight=-1.548495603643951, enabled=False)
	DefaultConnectionGene(key=(-5, 9), weight=0.08864438252079193, enabled=False)
	DefaultConnectionGene(key=(-4, 0), weight=-1.1194856962856854, enabled=True)
	DefaultConnectionGene(key=(-4, 3), weight=1.1746170098927435, enabled=True)
	DefaultConnectionGene(key=(-4, 6), weight=0.2954938271602887, enabled=True)
	DefaultConnectionGene(key=(-4, 7), weight=-0.42243881988730525, enabled=False)
	DefaultConnectionGene(key=(-4, 8), weight=-0.39533179857153933, enabled=False)
	DefaultConnectionGene(key=(-4, 9), weight=-1.5564069742415014, enabled=False)
	DefaultConnectionGene(key=(-4, 10), weight=-0.11199375355148677, enabled=True)
	DefaultConnectionGene(key=(-3, 1), weight=0.823819071594964, enabled=True)
	DefaultConnectionGene(key=(-3, 2), weight=0.3329807137433546, enabled=True)
	DefaultConnectionGene(key=(-3, 3), weight=2.6708869382950384, enabled=False)
	DefaultConnectionGene(key=(-3, 4), weight=-0.5566292188702973, enabled=True)
	DefaultConnectionGene(key=(-3, 9), weight=0.4651465300507318, enabled=False)
	DefaultConnectionGene(key=(-2, 0), weight=0.19755909917403866, enabled=False)
	DefaultConnectionGene(key=(-2, 3), weight=4.20084721561307, enabled=True)
	DefaultConnectionGene(key=(-2, 5), weight=-1.3092786153161597, enabled=False)
	DefaultConnectionGene(key=(-2, 7), weight=-1.4199579055400473, enabled=False)
	DefaultConnectionGene(key=(-2, 8), weight=0.6778982552014529, enabled=False)
	DefaultConnectionGene(key=(-2, 9), weight=0.8094812348200062, enabled=False)
	DefaultConnectionGene(key=(-2, 10), weight=0.3939657762275981, enabled=False)
	DefaultConnectionGene(key=(-1, 1), weight=-0.25741524574004737, enabled=False)
	DefaultConnectionGene(key=(-1, 4), weight=-0.9288865123942338, enabled=True)
	DefaultConnectionGene(key=(-1, 8), weight=1.488275798307087, enabled=False)
	DefaultConnectionGene(key=(-1, 9), weight=3.051775708593338, enabled=False)
	DefaultConnectionGene(key=(-1, 11), weight=0.47459105317976213, enabled=True)
	DefaultConnectionGene(key=(2, 2), weight=0.801662980618508, enabled=True)
	DefaultConnectionGene(key=(3, 3), weight=-0.3784654666310404, enabled=True)
	DefaultConnectionGene(key=(4, 4), weight=2.4519496588344327, enabled=True)
	DefaultConnectionGene(key=(7, 7), weight=-0.3523466945108795, enabled=False)
	DefaultConnectionGene(key=(8, 8), weight=-4.366920933570428, enabled=True)
	DefaultConnectionGene(key=(30, 0), weight=4.08515106029471, enabled=False)
	DefaultConnectionGene(key=(209, 3), weight=0.9030175692182629, enabled=False)
	DefaultConnectionGene(key=(260, 0), weight=-1.3012571953494865, enabled=False)
	DefaultConnectionGene(key=(652, 4), weight=0.11843361001968966, enabled=False)
	DefaultConnectionGene(key=(879, 1), weight=-3.586862548770805, enabled=False)
	DefaultConnectionGene(key=(1294, 3), weight=-0.37974104151001664, enabled=False)
	DefaultConnectionGene(key=(1352, 3), weight=0.17513292629516666, enabled=False)
	DefaultConnectionGene(key=(1402, 5), weight=-0.011403552616170376, enabled=False)
'''

nodes = extract_info(texto)

# Mostra as conexes de cada n
for key, node in nodes.items():
    print(f"Node {key} connections:")
    for connection in node['connections']:
        from_node, to_node = connection[0]
        weight = connection[1]['weight']
        print(f"  From: {from_node}, To: {to_node}, Weight: {weight}")
```
